## **1. Probability Basics**

---

## 🧠 1. What Is Probability?

> **Probability** quantifies uncertainty — it tells us how **likely** an event is to occur.

$$
\text{Probability of Event A} = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}
$$

* Always between **0 and 1**

  * 0 → impossible
  * 1 → certain

---

## 🎲 2. Key Terminologies

| Term                 | Meaning                                                |
| -------------------- | ------------------------------------------------------ |
| **Experiment**       | An action with uncertain outcomes (e.g. rolling a die) |
| **Sample Space (S)** | Set of all possible outcomes                           |
| **Event (E)**        | A subset of the sample space                           |
| **Outcome**          | A possible result of an experiment                     |

---

## 📏 3. Types of Probability

### a) **Theoretical Probability**

$$
P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total outcomes}}
$$

Example: Rolling a 3 on a fair die → $P(3) = \frac{1}{6}$

---

### b) **Experimental (Empirical) Probability**

$$
P(A) = \frac{\text{Number of times A occurs}}{\text{Number of trials}}
$$

Used when actual experiments or simulations are run.

---

### c) **Axiomatic Probability**

Follows 3 axioms:

1. $0 \leq P(A) \leq 1$
2. $P(S) = 1$
3. $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

## 🔍 **Bayes' Theorem**

---

### 📘 **What is Bayes' Theorem?**

Bayes’ Theorem helps us **update our beliefs** (probabilities) when we get new data or evidence.

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Where:

* $P(A|B)$: **Posterior** – probability of A after observing B
* $P(B|A)$: **Likelihood** – probability of B given A
* $P(A)$: **Prior** – initial belief about A
* $P(B)$: **Evidence** – probability of observing B

---

### 🧠 **Conceptual Example**

> 🧪 Suppose a disease affects 1% of a population.
> A test detects the disease 99% of the time (true positive rate), but also gives a false positive 5% of the time.
> You test **positive**. What is the probability you **actually have the disease**?

Let's break it down:

* $P(Disease) = 0.01$
* $P(NoDisease) = 0.99$
* $P(Positive|Disease) = 0.99$
* $P(Positive|NoDisease) = 0.05$

#### Using Bayes' Theorem:

$$
P(Disease|Positive) = \frac{P(Positive|Disease) \cdot P(Disease)}{P(Positive)}
$$

$$
P(Positive) = (0.99)(0.01) + (0.05)(0.99) = 0.0099 + 0.0495 = 0.0594
$$

$$
P(Disease|Positive) = \frac{0.0099}{0.0594} ≈ 0.1666
$$

➡️ Even after testing **positive**, your chance of having the disease is only **16.66%**!

---

### 🤯 Why Is This Important in AI?

| Use Case                   | Explanation                                        |
| -------------------------- | -------------------------------------------------- |
| **Naive Bayes Algorithm**  | Spam detection, sentiment analysis, classification |
| **Recommendation Systems** | Updating preferences based on past behavior        |
| **Medical Diagnosis**      | Probabilistic reasoning under uncertainty          |
| **Bayesian Networks**      | Structured probabilistic models                    |
| **Anomaly Detection**      | If a new event has a very low posterior, flag it   |

---

### 🧮 Python Example

```python
# Bayes' Theorem Implementation

def bayes(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
    return (p_b_given_a * p_a) / p_b

# Example: disease test
p_a = 0.01  # Prior: P(Disease)
p_b_given_a = 0.99  # Likelihood: P(Positive | Disease)
p_b_given_not_a = 0.05  # False positive rate: P(Positive | No Disease)

posterior = bayes(p_a, p_b_given_a, p_b_given_not_a)
print(f"Probability of having disease given a positive test: {posterior:.2%}")
```

## 🎲 **2. Random Variables – Explained from Scratch**

---

### 📌 **What is a Random Variable?**

A **random variable** is a variable whose value is the outcome of a random process or experiment.

There are **2 types**:

| Type           | Description                              | Example                           |
| -------------- | ---------------------------------------- | --------------------------------- |
| **Discrete**   | Takes **finite or countable** values     | Number of heads in 3 coin tosses  |
| **Continuous** | Takes **infinite** values within a range | Height, weight, time, temperature |

---

### 🔢 **Notation**

* Random variables are often denoted by **capital letters**: $X, Y, Z$
* Their possible values are written as: $X = x$

For example:

> Let $X$ be the outcome of rolling a die.
> Then $P(X = 3) = \frac{1}{6}$

---

### 📊 **Probability Mass Function (PMF)** – For Discrete

PMF gives the probability of each discrete outcome.

If $X$ is the number of heads in 2 coin flips:

| x        | 0    | 1   | 2    |
| -------- | ---- | --- | ---- |
| P(X = x) | 0.25 | 0.5 | 0.25 |

✔️ PMF satisfies:

* $P(X = x) \geq 0$
* $\sum P(X = x) = 1$

---

### 📈 **Probability Density Function (PDF)** – For Continuous

For a continuous random variable like height, we define a **PDF**, $f(x)$, such that:

$$
P(a \leq X \leq b) = \int_{a}^{b} f(x) dx
$$

We can't assign a single value probability $P(X = x)$; it's always 0 for continuous variables. Instead, we look at **intervals**.

---

### 🧠 **Expected Value and Variance**

For a discrete random variable $X$ with PMF $p(x)$:

* **Expected value (Mean):**

$$
E[X] = \sum x \cdot P(X = x)
$$

* **Variance:**

$$
Var(X) = E[X^2] - (E[X])^2
$$

These are foundational for ML models — expected value helps define average behavior, variance shows uncertainty.

---

### 🤖 **Why Are Random Variables Important in AI?**

| AI Area                | Use of Random Variables                                   |
| ---------------------- | --------------------------------------------------------- |
| **Bayesian Inference** | Posterior distributions are defined over random variables |
| **Markov Chains**      | States are modeled as random variables                    |
| **Probabilistic ML**   | Randomness is introduced in models like Naive Bayes, HMMs |
| **Generative Models**  | GANs, VAEs use random variables to sample outputs         |

---

### 🧮 Python Example: Simulating a Discrete Random Variable

```python
import numpy as np

# Toss a fair coin 1000 times
outcomes = np.random.choice([0, 1], size=1000)  # 0 = tails, 1 = heads

# Estimate probability
heads_prob = np.sum(outcomes) / len(outcomes)
print(f"Estimated Probability of Heads: {heads_prob:.2f}")
```

## **3. Probability Distributions: Normal & Binomial**

---

### 🔹 What Is a Probability Distribution?

It’s a function that describes **how the values of a random variable are distributed** — i.e., what values are likely to occur and how frequently.

---

## 1️⃣ **Binomial Distribution** – *For Discrete Outcomes*

---

### 📌 Use When:

* You have **n independent trials**
* Each trial has **2 outcomes**: success (p), failure (1-p)
* You want to know the **number of successes**

### 📘 Formula:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Where:

* $n$: number of trials
* $k$: number of successes
* $p$: probability of success on a single trial

---

### 🧠 Example:

> Toss a coin 5 times. What is the probability of getting exactly 3 heads?

$$
n = 5, k = 3, p = 0.5
\Rightarrow P(X=3) = \binom{5}{3} (0.5)^3 (0.5)^2 = 0.3125
$$

---

### 🔧 Python Simulation:

```python
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.5
x = range(n+1)
probs = binom.pmf(x, n, p)

plt.bar(x, probs)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

---

## 2️⃣ **Normal Distribution** – *For Continuous Data*

---

### 📌 Use When:

* The data is **continuous**
* The distribution is **symmetric and bell-shaped**
* Many real-world phenomena follow this (height, test scores)

---

### 📘 Formula (PDF):

$$
f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

Where:

* $\mu$: mean
* $\sigma^2$: variance

---

### 🧠 Properties:

* Mean = Median = Mode
* 68% of data lies within 1 standard deviation
* 95% within 2, 99.7% within 3 (**Empirical Rule**)

---

### 🔧 Python Example:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
mean, std = 0, 1
y = norm.pdf(x, mean, std)

plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid()
plt.show()
```

## **4. Mean, Variance & Expectation**

---

### 🔹 1. **Mean (μ)** — *Measure of Central Tendency*

#### 📘 Definition:

The average of a random variable's possible values.

#### 📌 Formula (Discrete):

$$
\mu = \frac{1}{n} \sum_{i=1}^n x_i
$$

Or for a **probability distribution**:

$$
\mu = \sum_{i} x_i \cdot P(x_i)
$$

#### 🧠 Example:

If scores = \[2, 4, 6, 8],

$$
\text{Mean} = \frac{2 + 4 + 6 + 8}{4} = 5
$$

---

### 🔹 2. **Expectation (E\[X])** — *Weighted Mean of Random Variable*

#### 📘 Definition:

Expected value is the **long-run average outcome**.

#### 📌 Formula (Discrete):

$$
E[X] = \sum x_i \cdot P(x_i)
$$

#### 🧠 Example:

You roll a fair die:

$$
E[X] = \sum_{x=1}^{6} x \cdot \frac{1}{6} = \frac{21}{6} = 3.5
$$

> Expectation is just the **mean in probabilistic terms**.

---

### 🔹 3. **Variance (σ²)** — *Measure of Spread*

#### 📘 Definition:

How far values lie from the mean.

#### 📌 Formula:

$$
Var(X) = \sum P(x_i) (x_i - \mu)^2
$$

Or for sample data:

$$
s^2 = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})^2
$$

#### 🧠 Example:

If X = \[2, 4, 6], Mean = 4:

$$
Var = \frac{(2-4)^2 + (4-4)^2 + (6-4)^2}{3} = \frac{4 + 0 + 4}{3} = 2.67
$$

---

### 🔹 4. **Standard Deviation (σ)**

Just the square root of variance.

$$
\sigma = \sqrt{Var(X)}
$$

---

| --------------- | -------------------------------------------------------------- |
| **Mean**        | Centering data, loss functions                                 |
| **Variance**    | Feature scaling, PCA                                           |
| **Expectation** | Expected reward (Reinforcement Learning), probabilistic models |

## 🎯 **Conditional Probability** – Full Guide with Videos & Examples

---

### 🔹 What is Conditional Probability?

It’s the probability of an event **A** happening **given** that another event **B** has already occurred.

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

* **A ∩ B** → both A and B happen
* **P(B)** ≠ 0 (since we’re assuming B happened)

---

### 🔹 Intuitive Example:

From a deck of 52 cards:

**Q:** What’s the probability that a card is a King **given** it is a face card?

* Face cards = {Jack, Queen, King} × 4 suits = **12**
* Kings = 4

$$
P(King \mid Face) = \frac{4}{12} = \frac{1}{3}
$$

---

### 🔹 Real-Life AI Example:

In spam detection:

* **A:** Email contains "Buy Now"
* **B:** Email is spam

You compute:

$$
P(A \mid B) = \text{"Probability of seeing 'Buy Now' given the email is spam"}
$$

Used in **Naive Bayes classifier**!

---

### 🔹 Python Example:

```python
# Example: Given 30 people drink tea, 18 drink coffee, 12 drink both.
P_tea = 30/50
P_coffee = 18/50
P_both = 12/50

# Conditional Probability of coffee given tea
P_coffee_given_tea = P_both / P_tea
print(P_coffee_given_tea)  # Output: 0.4
```

---

### 🔸 Relationship with Bayes' Theorem:

$$
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
$$

Bayes’ theorem is built directly on conditional probability!

## 🎯 **6. Central Limit Theorem (CLT)**

---

### 📘 **What is the Central Limit Theorem?**

> The **Central Limit Theorem** states that:
>
> > If you take a large number of **random samples** from **any population** (with finite variance), the **distribution of sample means** will approach a **normal distribution**, **regardless of the shape of the original population**.

---

### 🔹 Formal Statement:

Let $X_1, X_2, ..., X_n$ be **i.i.d.** random variables with:

* Mean = μ
* Variance = σ²

Then as **n → ∞**, the sample mean:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

follows:

$$
\bar{X} \sim \mathcal{N}(\mu, \frac{\sigma^2}{n})
$$

---

### 🧠 Intuition:

* Suppose you have a weird-shaped distribution (skewed, uniform, etc.).
* You take **many small samples** (e.g., size = 30) and compute their **means**.
* The **distribution of these means** will look like a **bell curve (Normal Distribution)**!

📌 This is why we can **assume normality** in machine learning (even when original data isn't normal)!

---

### 📊 Visual Example:

Imagine measuring the average height of different groups (n=30) from a population with weird height distribution.

* **Single sample**: Random
* **Multiple sample means**: Approaches normal curve

> ✅ This makes **estimation, confidence intervals, and testing** reliable using z-scores and t-scores.
---

### 🧪 Python Code Example

```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2, size=10000)  # Skewed population

sample_means = []

for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title("Central Limit Theorem in Action")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
---

| Area                | Use                                              |
| ------------------- | ------------------------------------------------ |
| Model Evaluation    | Confidence intervals for metrics (e.g. accuracy) |
| A/B Testing         | Conversion rates follow CLT assumptions          |
| Bootstrapping       | Resampling and averaging                         |
| Gradient Estimation | Used in stochastic methods                       |

---

### 🧠 Summary:

| Term        | Meaning                                                        |
| ----------- | -------------------------------------------------------------- |
| CLT         | Sample means become normally distributed as n grows            |
| Required    | Independent, Identically Distributed (i.i.d.), Finite Variance |

````python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2, size=10000)  # Skewed population

sample_means = []

for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title("Central Limit Theorem in Action")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
````
