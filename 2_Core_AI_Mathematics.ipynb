# 📘 **1. Scalars, Vectors, Matrices, and Tensors**

---

## ✅ 1. **Scalar**

* A single number
* No direction, just **magnitude**
* Examples:

  * Temperature: $35^\circ C$
  * Loss value in AI: $0.002$

$$
\text{Scalar: } a = 7
$$

---

## ✅ 2. **Vector**

* An ordered array of numbers (1D)
* Has both **magnitude and direction**
* Can be 2D, 3D, or nD

$$
\text{Vector: } \vec{v} =
\begin{bmatrix}
3 \\
2
\end{bmatrix}
\quad \text{(2D vector)}
$$

### 🔹 In AI:

* Features of a data point
* Word embeddings in NLP
* Weights in neural networks

---

## ✅ 3. **Matrix**

* A 2D array of numbers
* Rows × Columns
* Can represent multiple vectors or data points

$$
\text{Matrix: }
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\quad \text{(3x2 matrix)}
$$

### 🔹 In AI:

* Input batches to models
* Image data (grayscale image = 2D matrix)
* Transformation matrices in deep learning

---

## ✅ 4. **Tensor**

* A generalization of scalars, vectors, and matrices
* Can have **any number of dimensions** (0D to nD)

### 🔹 In AI:

* RGB image: 3D tensor (width × height × color channels)
* Video: 4D tensor (frames × width × height × channels)
* Deep learning models handle everything as tensors

---


# 📘 **2. Matrix Operations**

Let’s use:

$$
A = \begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}, \quad
B = \begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

---

## ✅ 1. **Matrix Addition**

You can add matrices if they have the **same dimensions**. Just add element-wise:

$$
A + B =
\begin{bmatrix}
1+5 & 2+6 \\
3+7 & 4+8
\end{bmatrix}
=
\begin{bmatrix}
6 & 8 \\
10 & 12
\end{bmatrix}
$$

---

## ✅ 2. **Matrix Multiplication** (Dot Product)

To multiply $A \times B$, the number of **columns in A** must equal the number of **rows in B**.

$$
A \cdot B =
\begin{bmatrix}
(1×5 + 2×7) & (1×6 + 2×8) \\
(3×5 + 4×7) & (3×6 + 4×8)
\end{bmatrix}
=
\begin{bmatrix}
19 & 22 \\
43 & 50
\end{bmatrix}
$$

🔹 **Use in AI**: Dot products are core to forward propagation in neural networks.

---

## ✅ 3. **Matrix Transpose**

Flip the matrix over its diagonal:

$$
A^T =
\begin{bmatrix}
1 & 3 \\
2 & 4
\end{bmatrix}
$$

🔹 **Use in AI**: Needed for operations like backpropagation and aligning dimensions.

---

# 📘 **3.Identity and Inverse Matrices**

---

## ✅ 1. **Identity Matrix (I)**

* A **square matrix** with 1's on the diagonal and 0's elsewhere.
* Acts like **1** in multiplication.

$$
I_2 =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
\quad \text{(2×2 identity)}
$$

$$
I \cdot A = A \cdot I = A
$$

🔹 In AI: It’s used to initialize layers, apply transformations, or preserve values during multiplication.

---

## ✅ 2. **Inverse Matrix (A⁻¹)**

* The matrix that *undoes* the effect of A:

$$
A \cdot A^{-1} = A^{-1} \cdot A = I
$$

* Only **square matrices** with **non-zero determinant** have inverses.

---

### 🔍 Example:

$$
A = \begin{bmatrix}
4 & 7 \\
2 & 6
\end{bmatrix}
$$

The inverse is:

$$
A^{-1} = \frac{1}{(4*6 - 7*2)} \begin{bmatrix}
6 & -7 \\
-2 & 4
\end{bmatrix}
= \frac{1}{10}
\begin{bmatrix}
6 & -7 \\
-2 & 4
\end{bmatrix}
$$

So:

$$
A^{-1} = \begin{bmatrix}
0.6 & -0.7 \\
-0.2 & 0.4
\end{bmatrix}
$$

And:

$$
A \cdot A^{-1} = I
$$

---

## 🚫 Not All Matrices Have Inverses

* A matrix **without** an inverse is called **singular**.
* If **det(A) = 0**, then $A^{-1}$ doesn’t exist.

---

## 🧠 In AI and ML:

| Use               | Purpose                                     |
| ----------------- | ------------------------------------------- |
| Backpropagation   | Invert weight updates                       |
| Optimization      | Solve matrix equations                      |
| Linear Regression | Closed-form solution: $w = (X^TX)^{-1}X^Ty$ |

# 📘 **4.Determinants and Rank – Core Matrix Properties**

---

## ✅ 1. **Determinant (det)**

The **determinant** is a scalar value computed from a **square matrix**. It tells us:

* Whether a matrix is **invertible**
* How a matrix **scales** space (volume transformation)
* Whether the matrix has **linearly dependent rows or columns**

---

### 🔹 For 2×2 Matrix

$$
A = \begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\quad \Rightarrow \quad \text{det}(A) = ad - bc
$$

**Example:**

$$
A = \begin{bmatrix}
4 & 7 \\
2 & 6
\end{bmatrix}
\quad \Rightarrow \quad \text{det}(A) = (4×6) - (7×2) = 24 - 14 = 10
$$

✔️ **Non-zero** → Invertible
❌ **Zero** → Not invertible (singular matrix)

---

### 🔹 For 3×3 Matrix

Use the **rule of Sarrus** or **cofactor expansion**:

$$
A = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}
\quad \Rightarrow \quad
\text{det}(A) = a(ei − fh) − b(di − fg) + c(dh − eg)
$$

---

### 🧠 In AI/ML:

* Check if a solution exists for matrix equations
* Understand geometric transformations
* Used in closed-form solutions like linear regression

---

## ✅ 2. **Matrix Rank**

The **rank** of a matrix is the number of **linearly independent rows or columns**.

* Measures the **information or dimensions** contained
* Always ≤ min(number of rows, number of columns)

---

### 🔹 Examples:

$$
A = \begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\quad \Rightarrow \quad \text{Rank} = 2 \quad \text{(full rank)}
$$

$$
B = \begin{bmatrix}
1 & 2 \\
2 & 4
\end{bmatrix}
\quad \Rightarrow \quad \text{Rank} = 1 \quad \text{(row 2 = 2×row 1)}
$$

---

## 🔍 Summary Table

| Concept     | Definition                                 | AI Use                                    |
| ----------- | ------------------------------------------ | ----------------------------------------- |
| Determinant | Scalar representing matrix scale/inversion | Invertibility, geometry, system solutions |
| Rank        | Number of linearly independent rows/cols   | Data compression, PCA, feature redundancy |

# 📘 **5. Eigenvalues and Eigenvectors – Core Linear Algebra for AI**

---

## ✅ What Are They?

Given a **square matrix** $A$, an **eigenvector** $\vec{v}$ and corresponding **eigenvalue** $\lambda$ satisfy the equation:

$$
A \vec{v} = \lambda \vec{v}
$$

* You're scaling the vector $\vec{v}$ without changing its direction.
* $\lambda$ is how much the vector is stretched or shrunk.

---

## 🔍 Intuition

* Imagine $A$ as a **transformation** (like rotation, scaling).
* Most vectors will change direction when transformed by $A$.
* But **some special vectors** (eigenvectors) will **only stretch or compress**.
* The amount they stretch/shrink is the **eigenvalue**.

---

## ✏️ Simple Example:

Let’s say:

$$
A = \begin{bmatrix}
4 & -2 \\
1 & 1
\end{bmatrix}
$$

We solve the **characteristic equation**:

$$
\det(A - \lambda I) = 0
$$

This gives us **eigenvalues**, and from them we can solve for **eigenvectors**.

---

## 🧠 Why Important in AI?

| Application                            | Use                                                                  |
| -------------------------------------- | -------------------------------------------------------------------- |
| **PCA (Principal Component Analysis)** | Dimensionality reduction using eigenvectors of the covariance matrix |
| **Neural Networks**                    | Understanding weight dynamics                                        |
| **Graph Theory**                       | PageRank algorithm uses eigenvectors                                 |
| **Stability analysis**                 | In dynamical systems (like reinforcement learning)                   |

---

## 🎯 Summary

| Term                | Meaning                                               |
| ------------------- | ----------------------------------------------------- |
| **Eigenvalue (λ)**  | Amount of stretch/shrink                              |
| **Eigenvector (v)** | Vector that doesn’t change direction when transformed |
| **Used in**         | PCA, PageRank, stability of systems, deep learning    |

# 📘 **6. Linear Transformations and Projections**

---

## ✅ 1. **Linear Transformation**

A **linear transformation** is a function that **maps vectors to other vectors** in a way that preserves:

* **Vector addition**: $T(\vec{u} + \vec{v}) = T(\vec{u}) + T(\vec{v})$
* **Scalar multiplication**: $T(c\vec{v}) = cT(\vec{v})$

Most commonly represented as:

$$
T(\vec{x}) = A \cdot \vec{x}
$$

Where:

* $A$ is a **matrix**
* $\vec{x}$ is a **vector**

---

### 🔹 Examples of Linear Transformations:

| Transformation | Matrix                                                                              |
| -------------- | ----------------------------------------------------------------------------------- |
| Rotation       | $\begin{bmatrix} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{bmatrix}$ |
| Scaling        | $\begin{bmatrix} k & 0 \\ 0 & k \end{bmatrix}$                                      |
| Shearing       | $\begin{bmatrix} 1 & m \\ 0 & 1 \end{bmatrix}$                                      |
| Reflection     | $\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$ (over x-axis)                       |

🔁 Any of these matrices **transform space** by moving, rotating, or stretching vectors.

---

## 🧠 In AI & ML:

* **Neural networks** use transformations between layers.
* **Image transformations** (rotation, zoom, flip) are linear transformations.
* **Word embeddings** can be rotated or projected between languages using linear maps.

---

## ✅ 2. **Projections**

A **projection** is a specific kind of linear transformation that “drops” a vector onto a subspace (like a line or a plane).

---

### 🔹 Geometric Intuition:

* You shine a flashlight on a vector and its **shadow** falls on a subspace — that’s a **projection**.
* The projection is the **closest point** in the subspace to the original vector.

---

### 🔹 Projection onto a Line:

Given a vector $\vec{a}$ and you want to project another vector $\vec{b}$ onto $\vec{a}$:

$$
\text{proj}_{\vec{a}} \vec{b} = \frac{\vec{a} \cdot \vec{b}}{\vec{a} \cdot \vec{a}} \cdot \vec{a}
$$

---

### 📊 In AI:

| Use                      | Example                                         |
| ------------------------ | ----------------------------------------------- |
| Dimensionality Reduction | PCA projects high-D data onto fewer dimensions  |
| Data Visualization       | 3D → 2D projections (like t-SNE or UMAP)        |
| Feature Extraction       | Remove noise, keep principal features           |
| Vector Similarity        | Cosine similarity is based on projection length |

---

## 🧠 Visualization Summary

* **Linear transformation**: stretches, rotates, flips, scales
* **Projection**: flattens or reduces dimension by dropping onto a lower surface (line, plane, etc.)

# 📘 **1. Limits and Continuity – Core Calculus for AI**

---

## ✅ 1. **What is a Limit?**

A **limit** tells us the value that a function $f(x)$ approaches as $x$ gets **closer to a particular point**.

### 🔹 Notation:

$$
\lim_{x \to a} f(x) = L
$$

This means as $x$ gets close to $a$, the function value $f(x)$ gets close to $L$.

---

### 🔹 Simple Example:

$$
\lim_{x \to 2} (3x + 1) = 7
\quad \text{because} \quad 3(2) + 1 = 7
$$

Even without plugging in $x = 2$, the **behavior around** $x = 2$ gives the limit.

---

## ✅ 2. **One-sided Limits**

* Left-hand limit: $\lim_{x \to a^-} f(x)$
* Right-hand limit: $\lim_{x \to a^+} f(x)$

For the full limit to exist, both must be **equal**.

---

## ✅ 3. **What is Continuity?**

A function $f(x)$ is **continuous at $x = a$** if:

1. $\lim_{x \to a} f(x)$ exists
2. $f(a)$ is defined
3. $\lim_{x \to a} f(x) = f(a)$

---

### 🔹 Graphically:

* No **breaks**, **holes**, or **jumps**
* You can draw the graph **without lifting the pen**

---

### 🔹 Common Continuous Functions:

| Function Type      | Continuous?                        |
| ------------------ | ---------------------------------- |
| Linear             | ✅ Yes                              |
| Polynomial         | ✅ Yes                              |
| Trigonometric      | ✅ Yes (except at undefined points) |
| Rational functions | ❌ Only where denominator ≠ 0       |

---

## 🧠 Why Important in AI?

| Concept                  | Relevance                                               |
| ------------------------ | ------------------------------------------------------- |
| **Backpropagation**      | Uses limits to define **gradients**                     |
| **Activation Functions** | Must be **continuous and smooth** for optimization      |
| **Gradient Descent**     | Relies on function **slopes**, which require continuity |
| **Optimization**         | Discontinuous functions are hard to optimize            |

---

## 📊 Visual Intuition

* A **limit** shows how the function behaves **near** a point.
* **Continuity** ensures no **disruption** in value at that point.

# 📘 **2. Derivatives (Rules + Partial Derivatives)**

---

## ✅ 1. **What is a Derivative?**

The **derivative** of a function $f(x)$ at point $x$ tells us:

> **How fast** and in **what direction** $f(x)$ is changing as $x$ changes.

---

### 🔹 Formal Definition (Limit form):

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

* Measures **slope** or **rate of change**
* Graphically: **slope of the tangent line** at point $x$

---

## 🔧 2. **Basic Derivative Rules**

| Rule          | Formula                                                         | Example                          |
| ------------- | --------------------------------------------------------------- | -------------------------------- |
| Constant Rule | $\frac{d}{dx} c = 0$                                            | $\frac{d}{dx} 5 = 0$             |
| Power Rule    | $\frac{d}{dx} x^n = nx^{n-1}$                                   | $\frac{d}{dx} x^3 = 3x^2$        |
| Sum Rule      | $\frac{d}{dx} [f + g] = f' + g'$                                | $\frac{d}{dx}(x^2 + x) = 2x + 1$ |
| Product Rule  | $\frac{d}{dx} [fg] = f'g + fg'$                                 | $f(x)=x, g(x)=x^2$ → $x^2 + 2x$  |
| Quotient Rule | $\frac{d}{dx} \left[\frac{f}{g}\right] = \frac{f'g - fg'}{g^2}$ |                                  |
| Chain Rule    | $\frac{d}{dx} f(g(x)) = f'(g(x)) \cdot g'(x)$                   | $f(x) = \sin(x^2)$               |

---

## ✅ 3. **Partial Derivatives**

In AI, we often deal with functions of **multiple variables**, like:

$$
f(x, y) = x^2 + y^2
$$

To find how the function changes with respect to **one variable**, we take a **partial derivative**:

* $\frac{\partial f}{\partial x} = 2x$
* $\frac{\partial f}{\partial y} = 2y$

📌 Treat **other variables as constants** when differentiating.

---

### 🔹 Uses in AI/ML:

| Use Case             | Role of Derivatives                                 |
| -------------------- | --------------------------------------------------- |
| **Gradient Descent** | Uses **partial derivatives** to move toward minimum |
| **Backpropagation**  | Uses **chain rule** to compute error gradient       |
| **Optimization**     | Find maxima/minima of cost functions                |
| **Loss Functions**   | Derivatives give slope to reduce loss               |

---

## 🔍 Visualization:

* Imagine a **hill**:
  The **gradient vector** (collection of partial derivatives) points in the **steepest ascent** direction.
* In ML, we go in the **opposite direction** (steepest descent).

---

## 🔁 Quick Recap:

| Term                                       | Meaning                           |
| ------------------------------------------ | --------------------------------- |
| Derivative $f'(x)$                         | Slope at a point                  |
| Partial Derivative $\partial f/\partial x$ | Change w\.r.t one variable        |
| Chain Rule                                 | Derivative of composed functions  |
| Gradient                                   | Vector of all partial derivatives |

# 📘 **3. Chain Rule – The Heart of Backpropagation**

---

## ✅ 1. **What Is the Chain Rule?**

The **Chain Rule** helps compute the derivative of a **composition of functions**.

---

### 🔹 Formula:

If:

$$
y = f(g(x)) \quad \text{then} \quad \frac{dy}{dx} = f'(g(x)) \cdot g'(x)
$$

You differentiate the **outer function** and multiply it by the derivative of the **inner function**.

---

### 🔹 Example:

$$
y = \sin(x^2)
\Rightarrow \frac{dy}{dx} = \cos(x^2) \cdot 2x
$$

---

## ✅ 2. **Chain Rule in AI: Backpropagation**

In a neural network, each layer transforms the input:

$$
z = w \cdot x + b \quad (\text{Linear}) \\
a = \sigma(z) \quad (\text{Activation})
$$

To compute gradients like $\frac{d\text{Loss}}{dw}$, we **chain** together partial derivatives:

$$
\frac{dL}{dw} = \frac{dL}{da} \cdot \frac{da}{dz} \cdot \frac{dz}{dw}
$$

🔁 This is the **chain rule** at work!

---

## 🔄 Example in Backpropagation

### Assume:

* Output layer uses activation $a = \sigma(z)$
* Loss = Mean Squared Error $L = \frac{1}{2}(a - y)^2$

To update weights:

$$
\frac{dL}{dw} = \frac{dL}{da} \cdot \frac{da}{dz} \cdot \frac{dz}{dw}
$$

Where:

* $\frac{dL}{da} = a - y$
* $\frac{da}{dz} = \sigma'(z)$
* $\frac{dz}{dw} = x$

So,

$$
\frac{dL}{dw} = (a - y) \cdot \sigma'(z) \cdot x
$$

---

## 🎯 In Simple Terms:

> You apply the chain rule **layer by layer** from the **output layer to the input**, multiplying gradients as you go = **backpropagation**.

---

## 📊 Visual Example

Imagine a function chain:

```
Input x → z = w*x + b → a = ReLU(z) → Loss
```

You compute:

1. $\frac{d\text{Loss}}{da}$
2. $\frac{da}{dz}$
3. $\frac{dz}{dw}$

Then multiply all:

$$
\frac{d\text{Loss}}{dw} = \frac{d\text{Loss}}{da} \cdot \frac{da}{dz} \cdot \frac{dz}{dw}
$$

---

## 🧠 Key Insight for AI

| Step            | Purpose                                                      |
| --------------- | ------------------------------------------------------------ |
| Chain Rule      | Links derivatives through layers                             |
| Backpropagation | Uses chain rule to compute gradients                         |
| Training        | Updates weights using these gradients (via Gradient Descent) |

# 📘 **4. Gradient and Gradient Descent**

---

## ✅ 1. What is a **Gradient**?

A **gradient** is a vector that points in the direction of the **steepest increase** of a function.

For a function of multiple variables like:

$$
f(x, y) = x^2 + y^2
$$

The gradient is:

$$
\nabla f = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right] = [2x, 2y]
$$

🔹 Intuition:

* It's like standing on a hill — the gradient points **uphill**, and its **magnitude** tells you how steep it is.

---

## ✅ 2. Why is the Gradient Important in AI?

* In **machine learning**, we define a **loss function** (error between prediction and truth).
* The **goal** is to **minimize** this function by adjusting model weights.

But how?

👉 By moving in the **opposite direction of the gradient**:
**downhill = less error** = better model.

---

## ✅ 3. What is Gradient Descent?

**Gradient Descent** is an algorithm that:

1. Calculates the gradient of the loss function
2. Moves in the direction that **minimizes** the loss (negative gradient)
3. Updates the model parameters (like weights)

---

### 🔹 Formula (Single Variable):

$$
\theta := \theta - \eta \cdot \frac{dL}{d\theta}
$$

Where:

* $\theta$: parameter (e.g., weight)
* $\eta$: learning rate
* $\frac{dL}{d\theta}$: gradient of loss $L$

---

### 🔹 Formula (Multi-variable vector form):

$$
\vec{w} := \vec{w} - \eta \cdot \nabla L(\vec{w})
$$

* This updates all weights simultaneously using the gradient vector.

---

## 🔁 Example

Assume loss function:

$$
L(w) = (wx - y)^2
$$

Then:

$$
\frac{dL}{dw} = 2(wx - y) \cdot x
$$

Update rule:

$$
w := w - \eta \cdot 2(wx - y) \cdot x
$$

Repeat until loss is small ✅

---

## 🔥 Gradient Descent in AI & ML

| Application       | Description                                                     |
| ----------------- | --------------------------------------------------------------- |
| Neural Networks   | Train weights via backpropagation using gradients               |
| Linear Regression | Minimize MSE using gradients                                    |
| Deep Learning     | Massive gradient descent in high dimensions                     |
| Optimizers        | Variants like SGD, Adam, RMSProp improve basic gradient descent |

---

## 🎯 Challenges

| Problem               | Solution                                         |
| --------------------- | ------------------------------------------------ |
| Too slow              | Use momentum or adaptive optimizers              |
| Stuck in local minima | Use better initializations or stochastic methods |
| Overshooting          | Tune learning rate (η) properly                  |

---

## 📊 Visual Intuition

Imagine a **bowl-shaped loss surface**. The gradient is like the slope on the bowl.
You take **small steps** in the direction of steepest descent (opposite of gradient) to reach the bottom (minimum error).

---

## 🧠 Summary

| Concept           | Meaning                                  |
| ----------------- | ---------------------------------------- |
| Gradient          | Direction and rate of steepest increase  |
| Negative Gradient | Steepest descent – direction to minimize |
| Gradient Descent  | Optimization method using gradients      |
| Learning Rate     | Size of steps taken toward minimum       |

# 📘 **5. Integration – Basic Understanding for AI & ML**

---

## ✅ 1. What is Integration?

**Integration** is the **reverse process of differentiation**.
It helps calculate:

* **Area under a curve**
* **Total accumulated value**
* **Probabilities** in continuous distributions

---

### 🔹 Notation:

$$
\int f(x) \, dx
$$

This means “the integral of $f(x)$ with respect to $x$”.

---

## ✅ 2. Definite vs. Indefinite Integrals

| Type           | Meaning                                         | Example                                             |
| -------------- | ----------------------------------------------- | --------------------------------------------------- |
| **Indefinite** | General form without limits; gives function + C | $\int x^2 dx = \frac{x^3}{3} + C$                   |
| **Definite**   | Has upper/lower limits; gives area/value        | $\int_1^3 x dx = \frac{3^2}{2} - \frac{1^2}{2} = 4$ |

---

## ✅ 3. Basic Integration Rules

| Rule           | Formula                                                 |
| -------------- | ------------------------------------------------------- |
| Power Rule     | $\int x^n dx = \frac{x^{n+1}}{n+1} + C$ for $n \neq -1$ |
| Constant Rule  | $\int a dx = ax + C$                                    |
| Sum Rule       | $\int (f + g) dx = \int f dx + \int g dx$               |
| Exponential    | $\int e^x dx = e^x + C$                                 |
| Trig Functions | $\int \sin x dx = -\cos x + C$, etc.                    |

---

## ✅ 4. Visual Intuition

Imagine plotting a curve like $y = x^2$.
The area **under this curve from x = 1 to x = 3** is found by:

$$
\int_1^3 x^2 dx = \left[\frac{x^3}{3}\right]_1^3 = \frac{27 - 1}{3} = \frac{26}{3}
$$

🔹 **Integration = Accumulated value over an interval**

---

## ✅ 5. Why Integration Matters in AI

| Field                   | Role of Integration                                |
| ----------------------- | -------------------------------------------------- |
| **Probability**         | Areas under PDF curves (e.g., normal distribution) |
| **Bayesian ML**         | Marginal likelihoods involve integration           |
| **Regularization**      | Some penalty terms are defined using integrals     |
| **Physics-Inspired AI** | Lagrangians, energy minimization (integrals)       |

## ✅ 6. Link to Derivatives

> **Derivatives measure change. Integration measures total.**

They are **inverse operations**:

$$
\frac{d}{dx} \left( \int f(x) dx \right) = f(x)
$$
