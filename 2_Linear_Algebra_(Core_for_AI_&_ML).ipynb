# **1. Scalars, Vectors, and Matrices**

---

## 🟩 1. Scalars

### ✅ Definition:

A **scalar** is a single number. It represents **magnitude only**, with no direction.

📌 **Example:**

* $5$, $-3.2$, $\pi$

In AI:

* Scalars represent **single values** like loss, learning rate, or one data point.

---

## 🟦 2. Vectors

### ✅ Definition:

A **vector** is a list (array) of numbers — it has **both magnitude and direction**.

📌 **Example:**

$$
\vec{v} = \begin{bmatrix} 2 \\ -1 \\ 3 \end{bmatrix}
$$

### 🔹 Key Concepts:

* **Dimension (nD)**: A vector with $n$ elements is n-dimensional.
* **Vector addition:**

$$
\begin{bmatrix} 1 \\ 2 \end{bmatrix} + \begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 4 \\ 6 \end{bmatrix}
$$

* **Dot product:**

$$
\vec{a} \cdot \vec{b} = a_1b_1 + a_2b_2 + ... + a_nb_n
$$

In AI:

* Vectors represent **features of data**, **weights of models**, or **input/output values**.

---

## 🟨 3. Matrices

### ✅ Definition:

A **matrix** is a 2D array of numbers (rows and columns).

📌 **Example:**

$$
A = \begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

### 🔹 Key Operations:

* **Matrix addition** (element-wise)
* **Matrix multiplication** (row by column rule)
* **Transpose**: Flip rows & columns

  $$
  A^T = \begin{bmatrix}
  1 & 3 \\
  2 & 4
  \end{bmatrix}
  $$

In AI:

* Matrices are used for:

  * Representing datasets (rows = samples, columns = features)
  * Transformations in neural networks
  * Calculating gradients and outputs

---

## 🧠 Why This Is Crucial for AI

| Concept  | Used In                                 |
| -------- | --------------------------------------- |
| Scalars  | Loss value, learning rate               |
| Vectors  | Features, weights, gradients            |
| Matrices | Datasets, model layers, transformations |

# **2. Matrix Operations**
---

## 🟨 Matrix Operations – Complete Guide

### 🔹 1. **Matrix Addition & Subtraction**

You can **add or subtract** matrices of the **same dimensions** by adding/subtracting their corresponding elements.

📌 **Example:**

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

$$
A + B = \begin{bmatrix} 6 & 8 \\ 10 & 12 \end{bmatrix}, \quad A - B = \begin{bmatrix} -4 & -4 \\ -4 & -4 \end{bmatrix}
$$

---

### 🔹 2. **Scalar Multiplication**

Multiply **every element** of the matrix by a scalar value.

📌 Example:

$$
k = 2, \quad A = \begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix}
\Rightarrow kA = \begin{bmatrix} 2 & 6 \\ 4 & 8 \end{bmatrix}
$$

---

### 🔹 3. **Matrix Multiplication (Dot Product)**

For two matrices $A_{m \times n}$ and $B_{n \times p}$, the product $C_{m \times p} = A \cdot B$ is defined as:

$$
C_{i,j} = \sum_{k=1}^{n} A_{i,k} \cdot B_{k,j}
$$

📌 **Example:**

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
\Rightarrow AB = \begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}
$$

**Used in AI for:**

* Calculating outputs of layers
* Forward propagation
* Loss calculation

---

### 🔹 4. **Transpose of a Matrix**

Flip rows and columns:

$$
A^T_{i,j} = A_{j,i}
$$

📌 Example:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
\Rightarrow A^T = \begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix}
$$

---

### 🔹 5. **Identity Matrix (I)**

A square matrix with 1s on the diagonal, 0s elsewhere.

$$
I = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
\Rightarrow A \cdot I = A
$$

---

### 🔹 6. **Inverse Matrix**

For a square matrix $A$, if there's a matrix $A^{-1}$ such that:

$$
A \cdot A^{-1} = I
$$

Then $A^{-1}$ is the inverse of $A$.

Only square matrices with non-zero determinant have an inverse.

---

## 📌 Summary Table

| Operation       | Symbol   | Purpose              |
| --------------- | -------- | -------------------- |
| Addition        | $A + B$  | Combine data         |
| Subtraction     | $A - B$  | Difference           |
| Scalar Multiply | $kA$     | Scale data           |
| Matrix Multiply | $AB$     | Apply transformation |
| Transpose       | $A^T$    | Flip dimensions      |
| Identity        | $I$      | Neutral element      |
| Inverse         | $A^{-1}$ | Undo transformation  |

# **3. Systems of Linear Equations**

This is used heavily in **machine learning**, especially in:

* Solving optimization problems
* Linear regression
* Understanding how model parameters are fitted

---

## 🟢 What Is a System of Linear Equations?

It’s a set of equations that share the same variables.

### 📌 Example:

$$
\begin{cases}
2x + y = 8 \\
3x - y = 1
\end{cases}
$$

This system has **2 equations** and **2 variables**.

---

## 🔧 Goal:

To find values of $x$ and $y$ that satisfy **both equations** simultaneously.

---

## ✴️ Methods to Solve Systems

### 1. **Graphical Method**

* Plot the equations on a graph.
* The intersection point is the solution.
* Good for visual understanding but **not scalable**.

---

### 2. **Substitution Method**

* Solve one equation for a variable.
* Substitute in the other.

🔹 Example:
From $2x + y = 8 \Rightarrow y = 8 - 2x$
Plug into second:
$3x - (8 - 2x) = 1 \Rightarrow$ Solve for $x$

---

### 3. **Elimination Method**

* Add or subtract equations to eliminate one variable.

---

### 4. **Matrix Method (for ML/AI)**

If you have:

$$
\begin{cases}
2x + y = 8 \\
3x - y = 1
\end{cases}
\Rightarrow \text{Can be written as:}
$$

$$
AX = B \quad \text{where} \quad
A = \begin{bmatrix} 2 & 1 \\ 3 & -1 \end{bmatrix}, \quad
X = \begin{bmatrix} x \\ y \end{bmatrix}, \quad
B = \begin{bmatrix} 8 \\ 1 \end{bmatrix}
$$

### ✅ Solution:

$$
X = A^{-1}B
$$

This is **how AI algorithms solve systems of equations programmatically** using matrix inverse.

---

## 📌 Types of Solutions

| Condition          | Meaning                   |
| ------------------ | ------------------------- |
| Unique solution    | One point of intersection |
| No solution        | Lines are parallel        |
| Infinite solutions | Lines overlap completely  |

---

## 🧠 In AI/ML, You Use This When:

* Solving weights in linear regression.
* Solving simultaneous constraints in optimization.
* Working with **normal equations** in statistics.

# **4. Determinants and Inverse of a Matrix**

These are used in:

* Solving systems of linear equations
* Understanding matrix invertibility
* Machine learning model training (especially linear regression using normal equations)

---

## 🟦 1. **Determinant**

The **determinant** is a scalar value that can be computed from a **square matrix**.
It tells us **if the matrix is invertible** and gives insights about the matrix's transformation properties (like scaling or collapsing space).

---

### ✅ **Determinant of a 2x2 Matrix**

$$
A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}
\Rightarrow \text{det}(A) = ad - bc
$$

📌 Example:

$$
\text{det}\left(\begin{bmatrix} 3 & 4 \\ 2 & 5 \end{bmatrix}\right) = 3 \cdot 5 - 4 \cdot 2 = 15 - 8 = 7
$$

---

### ✅ **Determinant of a 3x3 Matrix**

$$
A = \begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix}
$$

Use the **cofactor expansion**:

$$
\text{det}(A) = a(ei - fh) - b(di - fg) + c(dh - eg)
$$

(For large matrices, Python or a calculator is recommended.)

---

### 🧠 **Interpretation in AI/ML:**

* If $\text{det}(A) = 0$, matrix is **singular** (non-invertible) → no unique solution exists.
* If $\text{det}(A) ≠ 0$, matrix is **invertible** → unique solution exists.

---

## 🟩 2. **Inverse of a Matrix**

If $A \cdot A^{-1} = I$, then $A^{-1}$ is the **inverse of matrix A**.

Only **square matrices** with **non-zero determinant** have an inverse.

---

### ✅ **2x2 Inverse Formula**

$$
A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}, \quad \text{if } \text{det}(A) ≠ 0,
$$

$$
A^{-1} = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$

📌 Example:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad \text{det}(A) = -2
$$

$$
A^{-1} = \frac{1}{-2} \begin{bmatrix} 4 & -2 \\ -3 & 1 \end{bmatrix} = \begin{bmatrix} -2 & 1 \\ 1.5 & -0.5 \end{bmatrix}
$$

---

## 📌 Inverse in AI/ML

Used to solve systems:

$$
AX = B \Rightarrow X = A^{-1}B
$$

Also used in:

* Linear regression (normal equation)
* Covariance matrix inversion (for Mahalanobis distance)
* Transformations in neural nets

---

## 💻 Python Example using NumPy

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
det = np.linalg.det(A)
print("Determinant:", det)

if det != 0:
    inv = np.linalg.inv(A)
    print("Inverse:\n", inv)
else:
    print("Matrix is singular, no inverse.")
```
---

## ✅ Summary

| Concept        | Key Formula                                             | Use in AI                         |
| -------------- | ------------------------------------------------------- | --------------------------------- |
| Determinant    | $ad - bc$ (for 2x2)                                     | Invertibility, volume scaling     |
| Inverse        | $A^{-1} = \frac{1}{\text{det}(A)} \times \text{adj}(A)$ | Solve $AX = B$, Linear regression |
| Non-invertible | $\text{det}(A) = 0$                                     | No unique solution                |

# **5. Eigenvalues and Eigenvectors**
---

These concepts are **central to deep learning, PCA (Principal Component Analysis)**, and many areas of **data science and AI**, such as:

* Dimensionality reduction
* Feature extraction
* Spectral clustering
* Stability analysis in neural networks

---

## 🧠 What Are Eigenvalues and Eigenvectors?

Given a **square matrix** $A$, an **eigenvector** $\vec{v}$ and **eigenvalue** $\lambda$ satisfy:

$$
A\vec{v} = \lambda \vec{v}
$$

### 📌 Interpretation:

The matrix transformation $A$ **stretches or compresses** the vector $\vec{v}$ by a factor $\lambda$ **without changing its direction**.

* $\vec{v}$: **Direction stays the same**
* $\lambda$: **Scaling factor**

---

## ✴️ Why Are They Important in AI?

| Application                            | Use                                                |
| -------------------------------------- | -------------------------------------------------- |
| **PCA (Principal Component Analysis)** | Reduce data dimensions while preserving variance   |
| **Neural Networks**                    | Understand weight updates, stability               |
| **Graph embeddings**                   | Spectral clustering uses eigenvectors of Laplacian |
| **Quantum computing**                  | State transitions modeled with eigenvectors        |

---

## 🔣 Example (2×2 Matrix)

Let’s say:

$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
$$

We find eigenvalues $\lambda$ by solving the **characteristic equation**:

$$
\text{det}(A - \lambda I) = 0
$$

$$
\Rightarrow \text{det}\left(\begin{bmatrix} 2 - \lambda & 1 \\ 1 & 2 - \lambda \end{bmatrix}\right) = 0
$$

$$
(2 - \lambda)^2 - 1 = 0 \Rightarrow \lambda = 1, 3
$$

Then plug back each $\lambda$ into:

$$
(A - \lambda I)\vec{v} = 0
$$

...to get the corresponding **eigenvectors**.

---

## 📐 Visualization

* **Eigenvectors** are directions in space (like axes) that don’t rotate under the matrix transformation.
* **Eigenvalues** tell how much the vector is stretched along those directions.

---

## 💻 Python Example (with NumPy)

```python
import numpy as np

A = np.array([[2, 1], [1, 2]])
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

---

## 📘 Summary Table

| Term                       | Meaning                                                     |
| -------------------------- | ----------------------------------------------------------- |
| **Eigenvector**            | A direction that doesn't change under matrix transformation |
| **Eigenvalue**             | A scalar that tells how much the eigenvector is stretched   |
| **Matrix Diagonalization** | Expressing matrix in terms of eigenvalues and eigenvectors  |
| **Principal Component**    | Eigenvector of covariance matrix with highest eigenvalue    |

# **6. Singular Value Decomposition (SVD)**
---
SVD is a **matrix factorization technique** used in:

* Dimensionality reduction
* Latent Semantic Analysis (LSA)
* Recommender systems (like Netflix, Amazon)
* Image compression
* Principal Component Analysis (PCA)

---

## 🧠 What is SVD?

For any **real $m \times n$ matrix $A$**, the SVD is:

$$
A = U \Sigma V^T
$$

Where:

* $U$: Orthogonal matrix of **left singular vectors** (size $m \times m$)
* $\Sigma$: Diagonal matrix of **singular values** (size $m \times n$)
* $V^T$: Transpose of an orthogonal matrix of **right singular vectors** (size $n \times n$)

> The singular values in $\Sigma$ are the **square roots of the eigenvalues** of $A^T A$ or $A A^T$.

---

## ✳️ Intuition

SVD breaks down a matrix into:

1. **Rotation (V^T)**
2. **Scaling/stretching (Σ)**
3. **Rotation again (U)**

It’s like saying:

> “Apply one rotation, scale/stretch the data, then rotate again.”

---

## 🧩 Example

Let:

$$
A = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}
$$

We can compute the SVD using NumPy:

```python
import numpy as np

A = np.array([[3, 1], [1, 3]])
U, S, VT = np.linalg.svd(A)

print("U:\n", U)
print("Singular Values:", S)
print("V^T:\n", VT)
```

---

## 🔬 Applications in AI

| Application                  | Purpose                                                            |
| ---------------------------- | ------------------------------------------------------------------ |
| **PCA**                      | Performs dimensionality reduction using SVD                        |
| **Recommender systems**      | Matrix completion via SVD                                          |
| **Image compression**        | Approximate image using top-k singular values                      |
| **Latent Semantic Analysis** | Extract topics from text corpora using SVD of term-document matrix |

---

## 📘 Summary Table

| Term                           | Meaning                                           |
| ------------------------------ | ------------------------------------------------- |
| **Singular values**            | Square roots of eigenvalues of $A^T A$            |
| **Left singular vectors (U)**  | Eigenvectors of $AA^T$                            |
| **Right singular vectors (V)** | Eigenvectors of $A^T A$                           |
| **Rank-k approximation**       | Keep top-k singular values to compress the matrix |

---

## 🖼️ Visual Concept

> Imagine a cloud of points → rotate it → stretch it → rotate again.
> That’s what SVD does to your data.
