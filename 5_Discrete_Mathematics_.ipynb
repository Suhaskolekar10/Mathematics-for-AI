## **1. Sets, Logic, and Propositional Logic**
---

## 📘 1. **Sets** – Basics

### 🔹 What is a Set?

A **set** is a collection of **distinct objects**.

```text
A = {1, 2, 3}       # A set with three elements
B = {2, 3, 4, 5}
```

### 🔹 Set Operations:

| Operation    | Symbol | Example                        | Result    |
| ------------ | ------ | ------------------------------ | --------- |
| Union        | A ∪ B  | {1,2,3} ∪ {2,3,4}              | {1,2,3,4} |
| Intersection | A ∩ B  | {1,2,3} ∩ {2,3,4}              | {2,3}     |
| Difference   | A - B  | {1,2,3} - {2,3}                | {1}       |
| Subset       | A ⊆ B  | {2,3} ⊆ {1,2,3}                | True      |
| Complement   | Aᶜ     | U - A (relative to universe U) | —         |

✅ **Used in**: clustering, classification labels, Venn diagrams.

---

## 📘 2. **Logic** – Foundation for Reasoning

### 🔹 Types of Logic:

* **Propositional Logic**: deals with simple true/false statements.
* **Predicate Logic**: introduces variables and quantifiers like ∀, ∃.

Let’s focus on **Propositional Logic** here.

---

## 📘 3. **Propositional Logic**

### 🔹 What is a Proposition?

A **proposition** is a statement that is either **True (T)** or **False (F)**.

Examples:

* “It is raining” → T or F
* “2 + 2 = 5” → F

---

### 🔹 Logical Operators:

| Symbol | Name    | Meaning        | Example    | Result |
| ------ | ------- | -------------- | ---------- | ------ |
| ¬      | NOT     | Negation       | ¬P (not P) | T→F    |
| ∧      | AND     | Conjunction    | P ∧ Q      | T∧F=F  |
| ∨      | OR      | Disjunction    | P ∨ Q      | T∨F=T  |
| →      | IMPLIES | Implication    | P → Q      | T→F=F  |
| ↔      | IFF     | If and only if | P ↔ Q      | T↔T=T  |

✅ Truth tables are used to understand these.

---

### 🧠 Truth Table Example (AND):

| P | Q | P ∧ Q |
| - | - | ----- |
| T | T | T     |
| T | F | F     |
| F | T | F     |
| F | F | F     |

---

### 🔹 Logical Equivalence

Two propositions are **logically equivalent** if they always have the same truth value.

Example:

$$
P \rightarrow Q \equiv \neg P \lor Q
$$

## **2. Functions and Relations**
---

### 🔹 1. **What is a Relation?**

A **relation** is a set of ordered pairs that link elements from one set (domain) to another (range).

> **Example:**
> If A = {1, 2, 3}, B = {4, 5},
> Then a **relation R** from A to B could be:
> **R = {(1, 4), (2, 5), (3, 4)}**

🧠 In AI: Relations help model **connections** (e.g., user-item interactions, graph structures, logic systems).

---

### 🔹 2. **What is a Function?**

A **function** is a special type of relation in which **each input has exactly one output**.

* Function: `f(x) = y`
* It maps each `x` from the domain to only one `y` in the range.

✅ Example:

```text
f(x) = x^2
f(2) = 4
```

---

### 🎯 Properties of Functions

| Property                   | Meaning                               |
| -------------------------- | ------------------------------------- |
| **One-to-One** (Injective) | Every output maps to one unique input |
| **Onto** (Surjective)      | Every element in codomain is mapped   |
| **Bijective**              | Both one-to-one and onto              |

---

### 🔁 Common Types of Functions

| Function      | Example           | Graph Shape     |
| ------------- | ----------------- | --------------- |
| Linear        | f(x) = 2x + 3     | Straight line   |
| Quadratic     | f(x) = x²         | Parabola        |
| Exponential   | f(x) = 2^x        | J-curve upward  |
| Logarithmic   | f(x) = log(x)     | Reverse of exp  |
| Piecewise     | Defined in parts  | Segmented lines |
| Step Function | floor(x), ceil(x) | Blocks/steps    |

✅ In machine learning, activation functions like **ReLU, sigmoid, softmax** are functions mapping real inputs to bounded outputs.

---

## 📊 Visual Understanding

A function is a **mapping**:

```
Domain         Codomain
  1  ─────▶     2
  2  ─────▶     4
  3  ─────▶     6
```

Here, `f(x) = 2x` is the function mapping input → output.

---

## 🧮 Function Notation & Composition

* **f(x)** means “function f applied to x”
* **Composite function**:
  If `f(x) = 2x` and `g(x) = x + 3`, then:

  $$
  f(g(x)) = 2(x + 3) = 2x + 6
  $$

---

    return x**2 + 2*x + 1

## **3. Graph Theory**
---
### 🔹 1. **What is a Graph?**

A **graph** is a collection of:

* **Nodes (Vertices)** – Entities
* **Edges (Links)** – Relationships between entities

Graphs can be:

* **Undirected** or **Directed**
* **Weighted** or **Unweighted**

> 📌 Example:
> Users = nodes, Friendships = undirected edges
> Websites = nodes, Clicks = directed edges

---

### 🔹 2. **Basic Terminology**

| Term                 | Meaning                                   |
| -------------------- | ----------------------------------------- |
| **Vertex (Node)**    | Fundamental unit/entity in the graph      |
| **Edge**             | Connection between two nodes              |
| **Degree**           | Number of edges connected to a node       |
| **Path**             | Sequence of nodes where each is connected |
| **Cycle**            | A closed loop of connected nodes          |
| **Connected Graph**  | A path exists between all nodes           |
| **Adjacency Matrix** | A matrix representation of connections    |

---

### 🔹 3. **Graph Types**

| Type                         | Description                        |
| ---------------------------- | ---------------------------------- |
| **Simple Graph**             | No loops or multiple edges         |
| **Directed Graph (Digraph)** | Arrows represent direction         |
| **Weighted Graph**           | Edges carry values (weights)       |
| **Bipartite Graph**          | Nodes split into two disjoint sets |
| **Tree**                     | A connected acyclic graph          |

---

## 🧠 Graphs in AI: Where Are They Used?

| Application Area          | Role of Graphs                             |
| ------------------------- | ------------------------------------------ |
| **Graph Neural Networks** | Nodes = data points, edges = relationships |
| **Social Networks**       | Friends, followers                         |
| **Recommender Systems**   | User-item relationships                    |
| **Knowledge Graphs**      | Entities and their connections             |
| **Biological Networks**   | Genes, proteins, neurons                   |

## 📊 Adjacency Matrix vs Adjacency List

* **Adjacency Matrix**: NxN matrix for N nodes
* **Adjacency List**: Dictionary or list storing neighbors

```python
# Adjacency List (Python)
graph = {
  "A": ["B", "C"],
  "B": ["A", "D"],
  "C": ["A", "D"],
  "D": ["B", "C"]
}
```
