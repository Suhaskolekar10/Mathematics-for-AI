# **1. Limits and Derivatives**
---

These are **foundational concepts** in calculus used in:

* **Gradient Descent**
* **Backpropagation** in Neural Networks
* **Cost/Objective Function Optimization**
* Understanding **change** in values (like error, weights, etc.)

---

## 🧠 1. What is a Limit?

A **limit** describes the behavior of a function **as it approaches a specific point**.

### ✳️ Example:

$$
\lim_{x \to 2} (3x + 1) = 7
$$

> This means as $x$ gets closer to 2, the function $3x + 1$ gets closer to 7.

---

## 📐 Visual Intuition

* A limit shows **what value a function is heading toward**, not necessarily what value it reaches.
* If a function is continuous at a point, the **limit equals the function value** there.

---

## 🧠 2. What is a Derivative?

A **derivative** measures **how fast a function is changing** at any point.

### Formula:

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

> It's the **slope of the tangent line** at point $x$.

---

### ✳️ Example:

If $f(x) = x^2$, then:

$$
f'(x) = 2x
$$

So at $x = 3$, the slope (rate of change) is $f'(3) = 6$.

---

## 📌 In Machine Learning:

| Concept                  | Derivative Involved                                      |
| ------------------------ | -------------------------------------------------------- |
| **Loss Function**        | Find where loss is minimum using gradient                |
| **Gradient Descent**     | Uses derivative to update weights                        |
| **Neural Networks**      | Backpropagation uses chain rule for multiple derivatives |
| **Activation Functions** | Sigmoid, ReLU, etc., require derivatives during training |

---

## 💻 Python Derivative Example

```python
import sympy as sp

x = sp.Symbol('x')
f = x**2
derivative = sp.diff(f, x)
print(derivative)  # Output: 2*x
```

---

## 📘 Summary Table

| Concept        | Meaning                                | Application                               |
| -------------- | -------------------------------------- | ----------------------------------------- |
| **Limit**      | Value function approaches near a point | Understand continuity, base of derivative |
| **Derivative** | Instantaneous rate of change           | Optimization in ML, slope, backprop       |
| **Gradient**   | Vector of partial derivatives          | Used in high-dimensional ML models        |

# **2. Rules of Differentiation**

When functions get complex (multiplication, division, nested functions), we apply **rules** to differentiate them.
Let’s break them down with clear **formulas, examples, and use cases in ML**.

---

## ✅ 1. **Power Rule**

If $f(x) = x^n$, then:

$$
f'(x) = n \cdot x^{n-1}
$$

🧠 **Use case:** Common when working with loss functions like Mean Squared Error $(x - y)^2$

---

## ✅ 2. **Sum and Difference Rule**

If $f(x) = u(x) \pm v(x)$, then:

$$
f'(x) = u'(x) \pm v'(x)
$$

Just differentiate each part separately.

---

## ✅ 3. **Product Rule**

If $f(x) = u(x) \cdot v(x)$, then:

$$
f'(x) = u'(x)v(x) + u(x)v'(x)
$$

### 🔹 Example:

Let $f(x) = x^2 \cdot \sin(x)$

$$
f'(x) = 2x \cdot \sin(x) + x^2 \cdot \cos(x)
$$

📌 **Used in ML:** When loss is multiplied with activation or weights.

---

## ✅ 4. **Quotient Rule**

If $f(x) = \frac{u(x)}{v(x)}$, then:

$$
f'(x) = \frac{u'(x)v(x) - u(x)v'(x)}{[v(x)]^2}
$$

### 🔹 Example:

Let $f(x) = \frac{x}{x^2 + 1}$

$$
f'(x) = \frac{(1)(x^2 + 1) - x(2x)}{(x^2 + 1)^2}
$$

---

## ✅ 5. **Chain Rule (Most Important!)**

If $f(x) = g(h(x))$, then:

$$
f'(x) = g'(h(x)) \cdot h'(x)
$$

> Used when functions are **nested** (think: sigmoid(ReLU(x² + 3x)))

### 🔹 Example:

Let $f(x) = \sin(x^2)$

$$
f'(x) = \cos(x^2) \cdot 2x
$$

📌 **Used in backpropagation** where layers are stacked.

---

## 🔬 Applications in AI/ML

| Rule          | Use Case                           |
| ------------- | ---------------------------------- |
| Power Rule    | Cost functions (MSE)               |
| Product Rule  | Loss \* weight cases               |
| Quotient Rule | Normalization, softmax derivatives |
| Chain Rule    | Backpropagation in neural networks |

---

## 🧠 Tip: Chain Rule in Backprop

Backpropagation = Applying the chain rule **repeatedly** layer by layer in a neural network:

$$
\frac{dL}{dW} = \frac{dL}{da} \cdot \frac{da}{dz} \cdot \frac{dz}{dW}
$$

---

## 💻 Python Example (SymPy)

```python
import sympy as sp

x = sp.Symbol('x')
f = sp.sin(x**2)
f_prime = sp.diff(f, x)
print(f_prime)
# Output: 2*x*cos(x**2)
```

---

## 📘 Summary Sheet

| Rule     | Formula                 | Trigger Condition       |
| -------- | ----------------------- | ----------------------- |
| Power    | $nx^{n-1}$              | Powers of $x$           |
| Product  | $u'v + uv'$             | Multiplying 2 functions |
| Quotient | $\frac{u'v - uv'}{v^2}$ | Dividing functions      |
| Chain    | $f'(g(x)) \cdot g'(x)$  | Nested functions        |

# **3. Partial Derivatives**
## 🔷 What Are Partial Derivatives?

When a function has **multiple variables** (e.g., $f(x, y)$), a **partial derivative** tells us **how the function changes as one variable changes**, keeping the others **constant**.

---

### 📌 Notation:

$$
\frac{\partial f}{\partial x} \quad \text{or} \quad f_x(x, y)
$$

This means: *how does $f(x, y)$ change when only $x$ changes, and $y$ stays constant?*

---

## 🧠 Why Partial Derivatives Matter in AI:

| Where?                | Why?                                                                 |
| --------------------- | -------------------------------------------------------------------- |
| **Gradient Descent**  | Each weight has its own partial derivative w\.r.t. the loss function |
| **Backpropagation**   | Uses chain rule with partials through network layers                 |
| **Optimization**      | Multivariable loss surfaces are navigated using partial gradients    |
| **Jacobian, Hessian** | Advanced use of partials in model analysis                           |

---

## 🔢 Example:

Let $f(x, y) = x^2y + 3xy^2$

### Partial derivative w\.r.t. $x$:

$$
\frac{\partial f}{\partial x} = 2xy + 3y^2
$$

### Partial derivative w\.r.t. $y$:

$$
\frac{\partial f}{\partial y} = x^2 + 6xy
$$

> ✅ Treat all other variables as **constants** while differentiating!

---

## 💡 Real-World AI Example

Let $L(w_1, w_2)$ be a **loss function** with 2 weights.

To update weights using **gradient descent**:

$$
w_1 = w_1 - \alpha \cdot \frac{\partial L}{\partial w_1}
$$

$$
w_2 = w_2 - \alpha \cdot \frac{\partial L}{\partial w_2}
$$

This is how **training** happens in deep learning.

---

## 💻 Python Example using SymPy

```python
import sympy as sp

x, y = sp.symbols('x y')
f = x**2 * y + 3 * x * y**2

# Partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("∂f/∂x:", df_dx)
print("∂f/∂y:", df_dy)
```

---

## 📘 Summary

| Concept                         | Meaning                                                     |
| ------------------------------- | ----------------------------------------------------------- |
| $\frac{\partial f}{\partial x}$ | Change in $f$ as $x$ changes, keeping other variables fixed |
| Gradient $\nabla f$             | Vector of all partial derivatives                           |
| Used in                         | Loss optimization, backpropagation, multivariable analysis  |

## **4. Gradients and Gradient Descent**

This is how neural networks **learn** and how models **minimize error**.

---

## 🧠 1. What is a Gradient?

A **gradient** is a **vector of partial derivatives**.

If you have a function $f(x, y)$, its gradient is:

$$
\nabla f = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right]
$$

* It points in the **direction of the steepest increase**.
* Its **negative** points in the direction of **steepest decrease** — which is what we want in AI to **minimize loss**.

---

## 📉 2. What is Gradient Descent?

Gradient Descent is an **optimization algorithm** to minimize functions (like loss functions in ML).

### Update Rule:

$$
\theta = \theta - \alpha \cdot \nabla_\theta J(\theta)
$$

Where:

* $\theta$: parameters (weights)
* $J(\theta)$: loss function
* $\alpha$: learning rate
* $\nabla_\theta J(\theta)$: gradient of loss w\.r.t. parameters

---

## ⚙️ 3. Real-Life Analogy

Imagine you're **blindfolded on a hill** and want to reach the **lowest point** (minimum error).
You feel the ground (compute gradient) and **take a small step downhill** (update weights).

Repeat this till you reach the bottom — that’s **gradient descent**.

---

## 🔢 4. Example

Let:

$$
f(x) = (x - 3)^2
$$

Gradient:

$$
f'(x) = 2(x - 3)
$$

Apply gradient descent:

```python
x = 0  # starting point
alpha = 0.1

for i in range(10):
    grad = 2 * (x - 3)
    x = x - alpha * grad
    print(f"Step {i+1}: x = {x:.4f}")
```

Output will show x approaching 3 (the minimum).

---

## 🧮 5. Why Gradient Descent in AI?

| Term             | Role                                        |
| ---------------- | ------------------------------------------- |
| Loss Function    | Measures model error                        |
| Gradient         | Shows how to change weights to reduce error |
| Gradient Descent | Updates weights to minimize loss            |

---

## 📦 6. Types of Gradient Descent

| Type                 | Description                        |
| -------------------- | ---------------------------------- |
| **Batch**            | Uses entire dataset for one update |
| **Stochastic (SGD)** | One data point per update          |
| **Mini-Batch**       | Small chunks of data (most common) |

---

## 📘 Summary Sheet

| Concept                    | Definition                                    |
| -------------------------- | --------------------------------------------- |
| **Gradient**               | Vector of partial derivatives                 |
| **Gradient Descent**       | Optimization algorithm using gradients        |
| **Learning Rate $\alpha$** | Size of each update step                      |
| **Loss Function**          | Tells how bad the model is                    |
| **Convergence**            | When further updates don’t improve model much |

# **5. basic understanding of Integrals**
---

## 🔷 1. What Is an Integral?

An **integral** is the **area under a curve**.

If differentiation is **slope of a curve**, integration is **area beneath the curve**.

---

## 🧠 2. Basic Terminology

| Term               | Meaning                              |
| ------------------ | ------------------------------------ |
| $\int f(x) dx$     | Indefinite integral (no limits)      |
| $\int_a^b f(x) dx$ | Definite integral (area from a to b) |
| Integrand          | Function being integrated            |
| dx                 | Variable of integration              |

---

## 🔢 3. Simple Examples

### **Indefinite Integral**

$$
\int x^2 \, dx = \frac{x^3}{3} + C
$$

✅ **No limits** → We add a constant $C$

---

### **Definite Integral**

$$
\int_0^2 x \, dx = \left[ \frac{x^2}{2} \right]_0^2 = \frac{4}{2} - 0 = 2
$$

✅ This gives a **numerical value** (area from x=0 to x=2 under the curve $y = x$)

---

## 📌 4. Why Integrals in AI?

| Use Case                  | Application                                                |
| ------------------------- | ---------------------------------------------------------- |
| **Probability**           | Area under PDF is found via integral                       |
| **Expected Values**       | Integral of value × probability                            |
| **Regularization**        | L2 norm involves integral of squared weights (continuous)  |
| **Bayesian Inference**    | Integrals over posterior distributions                     |
| **Gradient Optimization** | Some learning algorithms integrate loss over distributions |

---

## 💻 Python Example Using SymPy

```python
import sympy as sp

x = sp.symbols('x')
f = x**2

# Indefinite integral
indef = sp.integrate(f, x)
print("∫x² dx =", indef)

# Definite integral from 0 to 3
def_val = sp.integrate(f, (x, 0, 3))
print("∫₀³ x² dx =", def_val)
```

---

## 🧭 Summary

| Concept               | Integral Meaning                                |
| --------------------- | ----------------------------------------------- |
| $\int f(x) dx$        | Area under $f(x)$, no limits (general formula)  |
| $\int_a^b f(x) dx$    | Area between x = a and b                        |
| Derivative ↔ Integral | Inverse operations                              |
| Used in AI            | Probabilities, expectations, optimization, etc. |
