## 📌 **1. Numbers**

### 📘 **Types of Numbers**

| Type               | Description                                  | Examples        |
| ------------------ | -------------------------------------------- | --------------- |
| Natural Numbers    | Counting numbers starting from 1             | 1, 2, 3, 4...   |
| Whole Numbers      | Natural numbers + zero                       | 0, 1, 2, 3...   |
| Integers           | Positive and negative whole numbers          | -3, -2, 0, 1, 2 |
| Rational Numbers   | Numbers that can be expressed as fractions   | 1/2, 3/4, -5/2  |
| Irrational Numbers | Numbers that can't be expressed as fractions | π, √2, e        |

### 🔍 **Use in AI & Data Science**:

* Integer types are used to represent **classes** in classification problems.
* Real numbers are used for **model weights, probabilities, and gradients**.
* Data types (integers, floats) are crucial when performing **data wrangling or feature scaling**.

## 📌 **3. Order of Operations (BODMAS/PEMDAS)**

### 📘 **What is BODMAS?**

> It determines the order in which operations should be performed in a mathematical expression:

* **B**rackets
* **O**rders (Exponents, roots)
* **D**ivision
* **M**ultiplication
* **A**ddition
* **S**ubtraction

> Example:

```text
2 + 3 * (5^2 – 4) ÷ 2
= 2 + 3 * (25 – 4) ÷ 2
= 2 + 3 * 21 ÷ 2
= 2 + 63 ÷ 2
= 2 + 31.5
= 33.5
```

### 🔍 **Use in AI & Data Science**:

* Correct **evaluation of expressions** in Python code (important in NumPy/Pandas).
* Prevents **logic errors** in complex model equations and mathematical formulas.
* Crucial in **matrix arithmetic**, where improper order leads to incorrect results.

# **2. Variables and Expressions**
---

## 🧮 **1. What are Variables?**

A **variable** is a symbol (usually a letter) that **represents a number or value**. It can change or vary, which is why it’s called a *variable*.

### 🔹 Examples:

* $x = 5$
* $y = x + 2 \Rightarrow y = 7$

In math for AI, variables often represent:

* Features (e.g., $x_1, x_2$) in a dataset
* Weights (e.g., $w$) in neural networks
* Outputs (e.g., $y$, predictions)

---

## 🔢 **2. What are Expressions?**

An **expression** is a **combination of variables, numbers, and operations** (like +, −, ×, ÷) that represents a value.

### 🔹 Examples:

* $x + 3$
* $2a - b$
* $4x^2 + 7x + 5$

You **evaluate** an expression by plugging in values for the variables.

👉 Example:
If $x = 2$, then
$4x + 1 = 4(2) + 1 = 9$

---

## 🧠 **Why are These Important in AI?**

* **Expressions** become **equations** when used to model relationships (like regression: $y = mx + b$).
* Variables represent **data**, **parameters**, and **predictions** in machine learning.
* You’ll use variables in **loss functions**, **gradient calculations**, and **feature mappings**.

# **3. Fractions, Decimals, and Exponents**
---

## 🔸 1. **Fractions**

### ✅ Definition:

A **fraction** represents a part of a whole. It has two parts:

* **Numerator (top)** – how many parts
* **Denominator (bottom)** – total parts

📌 **Example:**

$$
\frac{3}{4} \text{ means 3 parts out of 4}
$$

### 🔹 Key Concepts:

* **Equivalent Fractions:**

  $$
  \frac{1}{2} = \frac{2}{4} = \frac{4}{8}
  $$
* **Adding/Subtracting:**

  $$
  \frac{1}{3} + \frac{1}{6} = \frac{2 + 1}{6} = \frac{3}{6} = \frac{1}{2}
  $$
* **Multiplying:**

  $$
  \frac{2}{3} \times \frac{3}{4} = \frac{6}{12} = \frac{1}{2}
  $$
* **Dividing:**
  Flip the second fraction and multiply

  $$
  \frac{2}{3} \div \frac{4}{5} = \frac{2}{3} \times \frac{5}{4} = \frac{10}{12} = \frac{5}{6}
  $$

---

## 🔸 2. **Decimals**

### ✅ Definition:

A **decimal** is another way to represent fractions or parts of a whole using powers of 10.

📌 **Example:**

$$
0.5 = \frac{1}{2}, \quad 0.75 = \frac{3}{4}
$$

### 🔹 Key Skills:

* Converting **fractions ↔ decimals**

  $$
  \frac{1}{4} = 0.25
  $$
* **Addition/Subtraction**: Line up the decimal point.
* **Multiplication/Division**: Shift decimal points as needed.

---

## 🔸 3. **Exponents (Powers)**

### ✅ Definition:

An **exponent** tells you how many times to multiply a number by itself.

📌 **Example:**

$$
2^3 = 2 \times 2 \times 2 = 8
$$

### 🔹 Rules of Exponents:

* $a^m \cdot a^n = a^{m+n}$
* $\frac{a^m}{a^n} = a^{m-n}$
* $(a^m)^n = a^{m \cdot n}$
* $a^0 = 1$
* $a^{-n} = \frac{1}{a^n}$

**Used in AI for:**

* Activations like softmax
* Weight decay (regularization)
* Loss functions with exponential behavior
