# 📘 **1. Integers, Fractions, Decimals, Percentages**

---

## ✅ **A. Integers**

### 🔹 **Definition:**

* Whole numbers that can be **positive, negative, or zero.**
* ℤ = {..., -3, -2, -1, 0, 1, 2, 3, ...}

### 🔹 **Types of Integers:**

* **Positive Integers**: {1, 2, 3, …}
* **Negative Integers**: {… -3, -2, -1}
* **Zero (0)**: Neither positive nor negative.

### 🔹 **Properties:**

* **Closure**: a + b is an integer
* **Associative**: (a + b) + c = a + (b + c)
* **Commutative**: a + b = b + a
* **Distributive**: a × (b + c) = a×b + a×c

### 🔹 **Important Rules:**

| Operation      | Same Sign                             | Different Sign                        |
| -------------- | ------------------------------------- | ------------------------------------- |
| Addition       | Add & keep the sign                   | Subtract & keep sign of bigger number |
| Subtraction    | Convert to addition: a – b = a + (–b) |                                       |
| Multiplication | (+×+ = +), (–×– = +)                  | (+×– = –) or (–×+ = –)                |
| Division       | Same rules as multiplication          |                                       |

---

## ✅ **B. Fractions**

### 🔹 **Definition:**

* A number expressed as **a/b**, where **a** = numerator, **b ≠ 0** = denominator.

### 🔹 **Types:**

* **Proper Fraction**: Numerator < Denominator (e.g., 2/5)
* **Improper Fraction**: Numerator ≥ Denominator (e.g., 7/4)
* **Mixed Number**: Integer + proper fraction (e.g., 1¾)

### 🔹 **Equivalent Fractions:**

* Fractions that represent the same value.
  Example: ½ = 2/4 = 3/6

### 🔹 **Operations:**

| Operation            | Rule                                                         |
| -------------------- | ------------------------------------------------------------ |
| Addition/Subtraction | Convert to same denominator, then operate                    |
| Multiplication       | Multiply numerator × numerator and denominator × denominator |
| Division             | Flip the second fraction and multiply (Reciprocal rule)      |

### 🔹 **Simplification:**

* Divide numerator and denominator by their **HCF**

---

## ✅ **C. Decimals**

### 🔹 **Definition:**

* A number that includes a **decimal point** (e.g., 3.14, 0.5)

### 🔹 **Place Value Chart:**

| Units | Tenths | Hundredths | Thousandths |
| ----- | ------ | ---------- | ----------- |
| 5     | .4     | 0.04       | 0.004       |

### 🔹 **Converting:**

* **Fraction → Decimal**: Divide numerator by denominator.
* **Decimal → Fraction**: Write over 10, 100, etc., and simplify.

### 🔹 **Operations:**

* **Addition/Subtraction**: Line up decimal points
* **Multiplication**: Multiply normally, count total decimal places
* **Division**: Move decimals to make divisor whole → divide

---

## ✅ **D. Percentages**

### 🔹 **Definition:**

* A number expressed as **“per 100”**
* 1% = 1/100 = 0.01

### 🔹 **Conversions:**

| From     | To %        | Example    |
| -------- | ----------- | ---------- |
| Decimal  | ×100        | 0.75 = 75% |
| Fraction | (a/b) × 100 | ½ = 50%    |

### 🔹 **Finding:**

* **x% of y** = (x/100) × y

### 🔹 **Increase/Decrease:**

* **% Increase** = (Increase / Original) × 100
* **% Decrease** = (Decrease / Original) × 100

### 🔹 **Successive Percentage Change:**

* Use the formula:
  Total % change = a + b + (ab/100)

  > Example: 20% increase followed by 10% decrease =
  > 20 + (–10) + (20×–10)/100 = 10 – 2 = **8% increase**

---

## 🧠 Quick Tricks and Tips:

| Task              | Trick                                          |
| ----------------- | ---------------------------------------------- |
| 10% of a number   | Move decimal 1 place left                      |
| 5% of a number    | Half of 10%                                    |
| 1% of a number    | Move decimal 2 places left                     |
| Multiplying by 11 | For 2-digit numbers: 23 × 11 = 2 (2+3) 3 = 253 |

---

## 📌 Real-Life Use in AI:

* **Percentages** are used in accuracy, precision, recall.
* **Fractions and decimals** are key in data normalization.
* **Integers** are essential in indexing, loops, and algorithm steps.

# 📘 **2. Factors, Multiples, LCM, HCF**

---

## ✅ A. **Factors**

### 🔹 Definition:

A **factor** of a number divides it exactly without leaving any remainder.

> 📌 Example: Factors of 12 = 1, 2, 3, 4, 6, 12

### 🔹 Properties:

* Every number has at least two factors: **1 and itself**.
* **1 is a factor** of every number.
* **Number of factors** increases with composite numbers.

---

## ✅ B. **Multiples**

### 🔹 Definition:

A **multiple** of a number is the result of multiplying it by an integer.

> 📌 Example: Multiples of 4 = 4, 8, 12, 16, 20, ...

### 🔹 Properties:

* Every number is a multiple of itself.
* Every number has **infinite multiples**.
* **0 is a multiple** of every number.

---

## ✅ C. **Even, Odd, Prime, and Composite**

| Type      | Description                      | Example        |
| --------- | -------------------------------- | -------------- |
| Even      | Divisible by 2                   | 2, 4, 6, 8     |
| Odd       | Not divisible by 2               | 1, 3, 5, 7     |
| Prime     | Exactly 2 factors (1 and itself) | 2, 3, 5, 7, 11 |
| Composite | More than 2 factors              | 4, 6, 8, 9, 10 |

---

## ✅ D. **Highest Common Factor (HCF / GCD)**

### 🔹 Definition:

The **greatest number** that divides **two or more numbers exactly**.

> 📌 Example:
> HCF of 18 and 24 = 6
> (18: 1, 2, 3, 6, 9, 18 & 24: 1, 2, 3, 4, 6, 8, 12, 24)

### 🔹 Methods to Find HCF:

#### 1. **Listing Method** (For small numbers)

* List all factors of both numbers and choose the greatest common one.

#### 2. **Prime Factorization**

* Break both numbers into prime factors.
* Multiply common prime factors.

> 📌 Example:
> 18 = 2 × 3 × 3
> 24 = 2 × 2 × 2 × 3
> HCF = 2 × 3 = **6**

#### 3. **Euclidean Algorithm** (Fastest)

```text
HCF(a, b):
If b = 0 → return a
Else → HCF(b, a mod b)
```

> Example: HCF(48, 18)
> → HCF(18, 48 mod 18 = 12)
> → HCF(12, 6)
> → HCF(6, 0) → Answer = 6

---

## ✅ E. **Lowest Common Multiple (LCM)**

### 🔹 Definition:

The **smallest number** that is **common multiple** of two or more numbers.

> 📌 Example:
> LCM of 4 and 6 = 12
> (4: 4, 8, 12, 16... & 6: 6, 12, 18...)

### 🔹 Methods to Find LCM:

#### 1. **Listing Multiples**

→ Small numbers only

#### 2. **Prime Factorization**

* Take **maximum power** of each prime factor

> 📌 Example:
> 12 = 2² × 3
> 18 = 2 × 3²
> LCM = 2² × 3² = 36

#### 3. **Formula with HCF**

> 📌 LCM × HCF = a × b
> → LCM = (a × b) / HCF

---

## 🔄 HCF and LCM Together – Key Concept:

| Operation        | Symbol           | Use                    |
| ---------------- | ---------------- | ---------------------- |
| **HCF (or GCD)** | ∩ (intersection) | Minimize repeated task |
| **LCM**          | ∪ (union)        | Synchronize tasks      |

---

## 🧠 Real-Life Use in AI & Programming:

* **LCM** is used in scheduling and batch sizes (e.g., GPUs, datasets).
* **HCF** is useful in reducing ratios and simplifying fractions in preprocessing.
* These concepts help in writing **efficient logic**, especially for **modular arithmetic**.

---

## 📌 Quick Tips:

| Task                                      | Shortcut             |
| ----------------------------------------- | -------------------- |
| LCM of two coprime numbers                | Multiply the numbers |
| LCM & HCF same for equal numbers          | LCM = HCF = Number   |
| HCF > 1 means numbers are **not coprime** |                      |

# 📘 **3. Exponents and Roots**

---

## ✅ A. **Exponents (Powers)**

### 🔹 **Definition:**

An **exponent** (or power) tells how many times a number (called the base) is multiplied by itself.

> 📌 Example:
> $2^4 = 2 × 2 × 2 × 2 = 16$

Here:

* **2** is the base
* **4** is the exponent
* $2^4$ is the **power**

---

## ✅ B. **Laws of Exponents (Must-Memorize)**

| Rule Name               | Formula                                        | Example                   |
| ----------------------- | ---------------------------------------------- | ------------------------- |
| **Product of Powers**   | $a^m × a^n = a^{m+n}$                          | $2^3 × 2^2 = 2^5 = 32$    |
| **Quotient of Powers**  | $a^m ÷ a^n = a^{m-n}$                          | $5^4 ÷ 5^2 = 5^2 = 25$    |
| **Power of a Power**    | $(a^m)^n = a^{mn}$                             | $(3^2)^3 = 3^6$           |
| **Power of a Product**  | $(ab)^n = a^n × b^n$                           | $(2×3)^2 = 2^2 × 3^2$     |
| **Power of a Quotient** | $\left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}$ | $(4/2)^2 = 4^2 / 2^2 = 4$ |
| **Zero Exponent Rule**  | $a^0 = 1$ (a ≠ 0)                              | $7^0 = 1$                 |
| **Negative Exponent**   | $a^{-n} = \frac{1}{a^n}$                       | $2^{-3} = 1/8$            |

---

## ✅ C. **Important Exponent Patterns**

| Expression           | Value           |
| -------------------- | --------------- |
| $2^1, 2^2, 2^3, 2^4$ | 2, 4, 8, 16     |
| $10^1, 10^2, 10^3$   | 10, 100, 1000   |
| $a^{-1}$             | $\frac{1}{a}$   |
| $a^{1/2}$            | $\sqrt{a}$      |
| $a^{1/3}$            | $\sqrt[3]{a}$   |
| $a^{m/n}$            | $\sqrt[n]{a^m}$ |

---

## ✅ D. **Roots (Square Roots, Cube Roots, nth Roots)**

### 🔹 **Square Root (√):**

The number which when multiplied by itself gives the original number.

> 📌 $\sqrt{16} = 4$, because $4^2 = 16$

### 🔹 **Cube Root (∛):**

The number which when multiplied by itself **three times** gives the original number.

> 📌 $\sqrt[3]{27} = 3$, because $3^3 = 27$

### 🔹 **General Rule:**

* $a^{1/n} = \sqrt[n]{a}$
* $a^{m/n} = \sqrt[n]{a^m} = (\sqrt[n]{a})^m$

> 📌 Example:
> $8^{2/3} = (\sqrt[3]{8})^2 = 2^2 = 4$

# 📘 **4. Divisibility Rules (1 to 20)**

---

## ✅ What is Divisibility?

A number **A** is divisible by number **B** if the **remainder is zero** when A is divided by B.

> 📌 Example: 42 is divisible by 6 because 42 ÷ 6 = 7 with no remainder.

# 📘 **1. Variables and Expressions**

---

## ✅ A. **What Are Variables?**

### 🔹 Definition:

A **variable** is a symbol (usually a letter) used to represent a number whose value can change.

> 📌 Examples:

* `x`, `y`, `z` are common variables
* In `x + 5 = 10`, the variable is `x`

---

## ✅ B. **What Are Constants?**

* A **constant** is a fixed number. It doesn’t change.

> 📌 Example: In `x + 5`, the **5** is a constant.

---

## ✅ C. **What Are Expressions?**

### 🔹 Definition:

An **expression** is a mathematical phrase that can contain:

* Variables
* Constants
* Operations (+, −, ×, ÷)

> 📌 Examples:

* $x + 2$
* $3a - 7b + 12$
* $5(x - 3)$

🧠 **Note**: Expressions do **not** have an equal sign (`=`).
They are different from **equations**.

---

## ✅ D. **Types of Algebraic Expressions**

| Type           | Example              | Description |
| -------------- | -------------------- | ----------- |
| **Monomial**   | $5x$, $-3a^2$        | One term    |
| **Binomial**   | $x + 2$, $3a - 4$    | Two terms   |
| **Trinomial**  | $x^2 + 3x + 2$       | Three terms |
| **Polynomial** | $x^3 - 4x^2 + x - 1$ | 1+ terms    |

---

## ✅ E. **Terms, Coefficients, and Like Terms**

### 🔸 Term:

A **term** is a part of the expression separated by + or −.

> 📌 $4x + 3y - 7$ has **3 terms**: `4x`, `3y`, `-7`

### 🔸 Coefficient:

The **number** multiplied with a variable.

> 📌 In `7x`, the coefficient is **7**

### 🔸 Like Terms:

Terms with the **same variable** and **same power**.

> 📌 $3x$ and $-5x$ are like terms
> 📌 $3x$ and $3x^2$ are **not** like terms

---

## ✅ F. **Basic Operations with Expressions**

### 🔹 1. **Addition and Subtraction**

> Combine like terms only.

**Example:**
$(3x + 5) + (2x - 7) = 5x - 2$

### 🔹 2. **Multiplication**

Multiply each term.

**Example:**
$2(x + 3) = 2x + 6$

### 🔹 3. **Division**

Distribute if needed.

**Example:**
$\frac{6x + 9}{3} = 2x + 3$

---

## ✅ G. **Substitution**

Replace variables with given values to evaluate the expression.

**Example:**
If $x = 2$, find $3x + 4$
→ $3(2) + 4 = 10$

---

## 🧠 Why Important for AI?

| Concept        | AI Relevance                                          |
| -------------- | ----------------------------------------------------- |
| Variables      | Used in models (e.g. `x₁`, `x₂` in linear regression) |
| Expressions    | Found in cost functions, loss functions               |
| Substitution   | Feature input/output computation                      |
| Simplification | Optimizing models & algorithms                        |

# 📘 **2. Linear Equations and Inequalities**

---

## ✅ A. **What is a Linear Equation?**

A **linear equation** is an algebraic equation in which **each term is either a constant or the product of a constant and a single variable** (degree = 1).

### 🔹 General Form:

**1 variable:**

$$
ax + b = 0
$$

**2 variables:**

$$
ax + by = c
$$

> 📌 Example:

* $2x + 5 = 0$
* $3x - 7 = 11$
* $4x + 2y = 8$

---

## ✅ B. **Solving Linear Equations (1 Variable)**

### Step-by-step Method:

1. Bring variables to one side and constants to the other.
2. Simplify using basic operations.

> 📌 Example:
> Solve: $2x + 3 = 11$
> → $2x = 11 - 3 = 8$
> → $x = 8 ÷ 2 = 4$

---

## ✅ C. **Solving Equations (2 Variables)**

Use methods like:

* **Substitution**
* **Elimination**
* **Graphical method**

> 📌 Example:

$$
x + y = 10 \quad \text{and} \quad x - y = 2
$$

Add both: $2x = 12 \Rightarrow x = 6$
Then: $6 + y = 10 \Rightarrow y = 4$

---

## ✅ D. **What is a Linear Inequality?**

A **linear inequality** compares two expressions using:

* $<$
* $>$
* $\leq$
* $\geq$

> 📌 Example:

* $x + 3 > 7$
* $2y - 5 \leq 9$

---

## ✅ E. **Solving Inequalities**

### Rules (like equations, with one twist):

| Operation                   | Rule                            |
| --------------------------- | ------------------------------- |
| Add/Subtract                | Same as linear equations        |
| Multiply/Divide by Positive | Same as normal                  |
| Multiply/Divide by Negative | 🔄 **Flip the inequality sign** |

> 📌 Example:
> Solve:
> $-2x > 6$
> → Divide by -2:
> $x < -3$ ✅ (Sign flips!)

---

## ✅ F. **Graphical Representation**

* Linear equations form **straight lines**.
* Inequalities form **half planes** (shaded areas).

> 📌 $y = 2x + 1$ → line
> 📌 $y > 2x + 1$ → shaded region **above** the line
> 📌 $y \leq 2x + 1$ → shaded region **below**, line included

---

## ✅ G. **Systems of Equations vs. Inequalities**

| Feature  | Equation    | Inequality      |
| -------- | ----------- | --------------- |
| Graph    | Line        | Region          |
| Solution | Exact point | Range of values |
| Example  | $x + y = 5$ | $x + y > 5$     |

---

## ✅ H. **Application in AI & Data Science**

| Area                          | Application                                |
| ----------------------------- | ------------------------------------------ |
| **Linear Regression**         | Fits a straight line (linear equation)     |
| **Support Vector Machines**   | Use linear inequalities for classification |
| **Constraint Optimization**   | Inequalities define boundaries             |
| **Feature Scaling/Filtering** | Check value ranges (like $x > 0$)          |

# 📘 **3. Quadratic Equations**

---

## ✅ A. **What Is a Quadratic Equation?**

A **quadratic equation** is a second-degree polynomial equation in one variable.

### 🔹 Standard Form:

$$
ax^2 + bx + c = 0 \quad (a \neq 0)
$$

> Where:

* `a`, `b`, and `c` are constants
* `x` is the variable
* `a ≠ 0` (otherwise it becomes linear)

---

## ✅ B. **Examples of Quadratic Equations**

* $x^2 + 5x + 6 = 0$
* $3x^2 - 7x + 2 = 0$
* $x^2 - 9 = 0$ (note: b = 0)

---

## ✅ C. **Graph of a Quadratic Equation**

* It forms a **parabola** (U-shaped curve).

  * If `a > 0` → Opens **upward**
  * If `a < 0` → Opens **downward**
* The **vertex** is the turning point (maximum or minimum).
* The **roots (solutions)** are the x-values where the parabola intersects the x-axis.

---

## ✅ D. **Roots / Solutions of a Quadratic Equation**

The values of `x` that satisfy $ax^2 + bx + c = 0$

---

## ✅ E. **Methods to Solve Quadratic Equations**

---

### ✳️ 1. **Factorization Method**

Break into two linear factors.

> 📌 Example:
> $x^2 + 5x + 6 = 0$
> → $(x + 2)(x + 3) = 0$
> → $x = -2, -3$

---

### ✳️ 2. **Quadratic Formula**

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

> 📌 Example:
> $2x^2 + 3x - 5 = 0$
> → $a=2, b=3, c=-5$
> → $x = \frac{-3 \pm \sqrt{(3)^2 - 4(2)(-5)}}{2(2)}$
> → $x = \frac{-3 \pm \sqrt{49}}{4} = \frac{-3 \pm 7}{4}$
> → $x = 1, -2.5$

---

### ✳️ 3. **Completing the Square**

Make it a perfect square trinomial.

> 📌 Example:
> $x^2 + 6x + 5 = 0$
> → $x^2 + 6x = -5$
> → Add $\left(\frac{6}{2}\right)^2 = 9$ on both sides
> → $x^2 + 6x + 9 = 4$
> → $(x + 3)^2 = 4 \Rightarrow x = -3 \pm 2 \Rightarrow x = -1, -5$

---

## ✅ F. **Discriminant (D)**

$$
D = b^2 - 4ac
$$

Helps to determine the nature of the roots:

| Discriminant | Roots Nature        |
| ------------ | ------------------- |
| $D > 0$      | Real and distinct   |
| $D = 0$      | Real and equal      |
| $D < 0$      | Complex (imaginary) |

---

## ✅ G. **Sum and Product of Roots**

For $ax^2 + bx + c = 0$:

* Sum: $\alpha + \beta = -\frac{b}{a}$
* Product: $\alpha \cdot \beta = \frac{c}{a}$

---

## ✅ H. **Applications in AI and ML**

| Field                    | Usage                                                |
| ------------------------ | ---------------------------------------------------- |
| **Optimization**         | Used in convex loss functions (e.g., MSE)            |
| **Gradient Descent**     | Often finds minimum of quadratic-like cost functions |
| **Physics-based Models** | Projectile motion, curve fitting                     |
| **Computer Vision**      | Parabolic edge detection, shape recognition          |

# 📘 **4. Functions and Their Graphs**

---

## ✅ A. **What Is a Function?**

A **function** is a rule that assigns **each input (x)** to **exactly one output (y)**.

### 📌 Example:

$$
f(x) = 2x + 3
$$

* If $x = 1$, then $f(1) = 5$

---

## ✅ B. **Function Notation**

Instead of writing $y = 2x + 3$, we write:

$$
f(x) = 2x + 3
$$

* $f(x)$ is read as "function of x"
* It's just another way to write the **output y**

---

## ✅ C. **Types of Functions and Their Graphs**

---

### ✳️ 1. **Linear Function**

#### 🔹 General Form:

$$
f(x) = mx + c
$$

* $m$: slope (how steep the line is)
* $c$: y-intercept (where it crosses y-axis)

#### 🔹 Example:

$$
f(x) = 2x + 1
$$

#### 🔹 Graph:

* A **straight line**
* **Increasing** if $m > 0$, **decreasing** if $m < 0$

#### 🔹 Use in AI:

* Basis for **Linear Regression**, simple prediction models

---

### ✳️ 2. **Quadratic Function**

#### 🔹 General Form:

$$
f(x) = ax^2 + bx + c
$$

#### 🔹 Example:

$$
f(x) = x^2 - 4
$$

#### 🔹 Graph:

* A **parabola**
* Opens **up** if $a > 0$, **down** if $a < 0$
* **Vertex** is the lowest/highest point

#### 🔹 Use in AI:

* Used in **loss functions** like MSE (Mean Squared Error)
* Also appears in optimization and physics-based models

---

### ✳️ 3. **Exponential Function**

#### 🔹 General Form:

$$
f(x) = a^x \quad (a > 0,\ a \neq 1)
$$

#### 🔹 Example:

$$
f(x) = 2^x
$$

#### 🔹 Graph:

* **Rapid growth** or **decay**
* Always **above x-axis**
* If base $a > 1$: exponential growth
* If $0 < a < 1$: exponential decay

#### 🔹 Use in AI:

* **Logistic regression**, **activation functions** (like ReLU, sigmoid), **learning curves**

---

## ✅ D. **Comparing Graph Shapes**

| Type        | Shape         | Growth Pattern         |
| ----------- | ------------- | ---------------------- |
| Linear      | Straight line | Constant rate          |
| Quadratic   | Parabola      | Slow → Fast → Slow     |
| Exponential | Rapid curve   | Very fast growth/decay |

---

## ✅ E. **Basic Transformations**

Let’s say $f(x) = x^2$:

| Transformation         | New Function   | Effect on Graph           |
| ---------------------- | -------------- | ------------------------- |
| Vertical Shift Up      | $f(x) + c$     | Moves graph up            |
| Horizontal Shift Right | $f(x - c)$     | Moves graph right         |
| Vertical Stretch       | $a \cdot f(x)$ | Makes graph steeper       |
| Reflection (flip)      | $-f(x)$        | Flips graph across x-axis |

---

## ✅ F. **How It Connects to AI**

| AI Topic             | Related Function                                                                                       |
| -------------------- | ------------------------------------------------------------------------------------------------------ |
| Linear Regression    | Linear function (straight line fit)                                                                    |
| Loss Functions (MSE) | Quadratic function (minimizing a parabola)                                                             |
| Learning Curves      | Exponential/logarithmic                                                                                |
| Neural Networks      | Activation functions (ReLU, sigmoid, tanh) often use **piecewise, exponential, and sigmoid functions** |

# 📘 **1. Cartesian Coordinates (x, y)**

---

## ✅ A. **What Are Cartesian Coordinates?**

A **Cartesian coordinate system** represents points on a plane using **two perpendicular number lines**:

* **X-axis** (horizontal)
* **Y-axis** (vertical)

Each point is written as an **ordered pair**:

$$
(x, y)
$$

---

## ✅ B. **Structure of the Cartesian Plane**

The axes divide the plane into **4 quadrants**:

| Quadrant | Sign of x | Sign of y | Example  |
| -------- | --------- | --------- | -------- |
| I        | +         | +         | (2, 3)   |
| II       | –         | +         | (–4, 5)  |
| III      | –         | –         | (–3, –2) |
| IV       | +         | –         | (6, –1)  |

> 📌 Origin = (0, 0) → where X and Y axes cross

---

## ✅ C. **Plotting Points**

To plot a point like (3, 2):

1. Move **3 units right** on the X-axis
2. Move **2 units up** on the Y-axis
3. Mark the point 🔵

> 📌 (–2, –4): 2 units left, 4 units down

---

## ✅ D. **Distance Between Two Points**

Use the **Distance Formula**:

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

> 📌 Example:
> Between A(1, 2) and B(4, 6):

$$
d = \sqrt{(4 - 1)^2 + (6 - 2)^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

---

## ✅ E. **Midpoint Formula**

$$
\text{Midpoint} = \left(\frac{x_1 + x_2}{2},\ \frac{y_1 + y_2}{2}\right)
$$

> 📌 Between (2, 6) and (4, 8) → Midpoint = (3, 7)

---

## ✅ F. **Slope of a Line (m)**

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

* Positive slope → goes up
* Negative slope → goes down
* Zero slope → horizontal
* Undefined → vertical

> 📌 Example: From (1, 2) to (3, 6):

$$
m = \frac{6 - 2}{3 - 1} = \frac{4}{2} = 2
$$

---

## ✅ G. **Equation of a Line**

$$
y = mx + c
$$

* `m` is the slope
* `c` is the y-intercept

> 📌 Example: Line with slope 2, intercept 3:
> $y = 2x + 3$

# 📘 **2. Distance & Midpoint Formula – Notes with Examples**

---

## ✅ A. **Distance Formula**

The **distance between two points** $A(x_1, y_1)$ and $B(x_2, y_2)$ in a Cartesian plane is:

$$
\text{Distance (d)} = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

> 📌 **Why?**
> This is derived from the **Pythagorean Theorem**:

* The change in x is one side (Δx)
* The change in y is the other side (Δy)
* Distance is the hypotenuse of the right triangle

---

### 🔹 Example 1:

Find the distance between $A(1, 2)$ and $B(4, 6)$

$$
d = \sqrt{(4 - 1)^2 + (6 - 2)^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

---

### 🔹 Example 2 (negative coordinates):

Find distance between $P(-3, -2)$ and $Q(2, 1)$

$$
d = \sqrt{(2 + 3)^2 + (1 + 2)^2} = \sqrt{25 + 9} = \sqrt{34} \approx 5.83
$$

---

## 🧠 Where It’s Used in AI:

| Area                          | Usage                                             |
| ----------------------------- | ------------------------------------------------- |
| **Clustering (K-Means)**      | Finding nearest centroid using Euclidean distance |
| **KNN (K-Nearest Neighbors)** | Classifying based on shortest distance            |
| **Anomaly Detection**         | Outlier = point far from normal clusters          |
| **Word Embeddings**           | Similarity via cosine or Euclidean distance       |

---

## ✅ B. **Midpoint Formula**

The **midpoint** between $A(x_1, y_1)$ and $B(x_2, y_2)$ is:

$$
\text{Midpoint (M)} = \left( \frac{x_1 + x_2}{2},\ \frac{y_1 + y_2}{2} \right)
$$

> 📌 Think of this as the **average of the x’s and y’s**.

---

### 🔹 Example 1:

Find the midpoint of $A(2, 6)$ and $B(4, 8)$:

$$
M = \left( \frac{2 + 4}{2}, \frac{6 + 8}{2} \right) = (3, 7)
$$

---

### 🔹 Example 2:

Find midpoint between $(–3, 7)$ and $(5, –1)$

$$
M = \left( \frac{-3 + 5}{2}, \frac{7 - 1}{2} \right) = \left( 1, 3 \right)
$$

---

## 🧠 Where It’s Used in AI:

| Area                                 | Usage                                          |
| ------------------------------------ | ---------------------------------------------- |
| **Vector operations**                | Used to find centroids of line segments        |
| **Image transformations**            | Center of objects or bounding boxes            |
| **3D graphics in AI/VR**             | Object rotation about midpoints                |
| **Bounding Box in Object Detection** | Calculating center point for labels or anchors |

# 📘 **3. Line Equations – Slope-Intercept Form**

---

## ✅ A. **What Is the Slope-Intercept Form?**

It is the most common way to express the equation of a straight line:

$$
\boxed{y = mx + c}
$$

Where:

* $y$: output (dependent variable)
* $x$: input (independent variable)
* $m$: **slope** of the line (rise/run)
* $c$: **y-intercept** (where the line crosses the y-axis)

---

## ✅ B. **Understanding the Slope (m)**

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

| Slope Type | Interpretation                  |
| ---------- | ------------------------------- |
| $m > 0$    | Increasing line ↗               |
| $m < 0$    | Decreasing line ↘               |
| $m = 0$    | Horizontal line →               |
| Undefined  | Vertical line ↑↓ (x = constant) |

---

## ✅ C. **Understanding the Y-Intercept (c)**

* It’s the point where the line cuts the **y-axis** (x = 0)
* Plug $x = 0$ into the equation to find $y = c$

---

## ✅ D. **Graphing a Line from Equation**

### Example:

$$
y = 2x + 1
$$

1. **Slope** = 2 → rise = 2, run = 1
2. **Y-Intercept** = 1 → plot (0, 1)
3. Use slope to plot next point: from (0,1) go up 2, right 1 → (1,3)

---

## ✅ E. **How to Form Equation of a Line**

### 🔹 Given slope $m$ and y-intercept $c$:

Just plug into $y = mx + c$

### 🔹 Given two points $(x_1, y_1), (x_2, y_2)$:

1. Find slope $m = \frac{y_2 - y_1}{x_2 - x_1}$
2. Use point-slope form:

$$
y - y_1 = m(x - x_1)
$$

3. Rearrange to get slope-intercept form.

---

### 🔹 Example: Through (2, 3) and (4, 7)

* Slope:

$$
m = \frac{7 - 3}{4 - 2} = \frac{4}{2} = 2
$$

* Use:

$$
y - 3 = 2(x - 2) \Rightarrow y = 2x - 1
$$

---

## ✅ F. Real-World Use in AI

| Concept               | Example                                   |
| --------------------- | ----------------------------------------- |
| **Linear Regression** | Finding best-fit line through data points |
| **Classification**    | Decision boundary in 2D (line)            |
| **Loss Functions**    | Parabola intersects linear model          |
| **Gradient Descent**  | Uses slope to minimize loss               |

---

## 🔁 Recap

* $y = mx + c$: Slope-intercept form
* $m = \frac{\Delta y}{\Delta x}$: Slope = rise/run
* $c$: Y-intercept
* Use point-slope to derive equation from two points

# 📘 **4. Understanding 2D and 3D Spaces in Math & AI**

---

## ✅ A. **What Is a Dimension?**

A **dimension** represents a measurable direction in space:

| Dimension | Description  | Example                 |
| --------- | ------------ | ----------------------- |
| 0D        | A point      | `.` (no length, width)  |
| 1D        | A line       | `────` (only x-axis)    |
| 2D        | A plane      | x & y → flat shapes     |
| 3D        | A volume     | x, y, z → objects       |
| nD        | Higher space | Data with many features |

---

## ✅ B. **2D Space (Two-Dimensional)**

* Axes: **x** (horizontal) and **y** (vertical)
* Each point is represented as:

  $$
  (x, y)
  $$

### 🔹 Real-World Examples:

| Use    | Example                       |
| ------ | ----------------------------- |
| Graphs | Plotting y = 2x + 1           |
| Images | 2D grid of pixels             |
| Data   | Scatter plots with 2 features |

### 🔹 AI Use Cases:

* Visualizing classification (e.g., spam vs non-spam)
* 2-feature clustering (e.g., age vs income)
* Decision boundary of simple models (e.g., linear classifier)

---

## ✅ C. **3D Space (Three-Dimensional)**

* Axes: **x**, **y**, and **z**
* Each point is:

  $$
  (x, y, z)
  $$

> You can think of z as **depth** (out of the screen)

### 🔹 3D Coordinates Visualization:

* (0, 0, 0) = Origin
* (2, 3, 4) → 2 units right (x), 3 up (y), 4 outwards (z)

---

### 🔹 Real-World Examples:

| Use             | Example                            |
| --------------- | ---------------------------------- |
| Geometry        | Cubes, spheres                     |
| Games & AR      | 3D environments                    |
| Physics         | Force vectors                      |
| Computer Vision | 3D object detection from 2D images |

---

### 🔹 AI Use Cases:

| Task                   | 3D Example                         |
| ---------------------- | ---------------------------------- |
| **3D CNNs**            | Video frame analysis               |
| **Point Clouds**       | Lidar in self-driving cars         |
| **Data Visualization** | PCA projections of 3+ feature data |
| **GANs & 3D Modeling** | AI-generated 3D objects            |

---

## ✅ D. **Distance in 3D Space**

Same as 2D, but with an extra z-axis term:

$$
\text{Distance} = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}
$$

---

## ✅ E. **Why This Matters in AI**

| Concept        | Why It’s Important                                      |
| -------------- | ------------------------------------------------------- |
| Feature Space  | Each axis = a feature (e.g., height, age, salary)       |
| Clustering     | High-dimensional spaces for customer segmentation       |
| Neural Nets    | Input tensors often live in 3D+                         |
| Data Reduction | PCA, t-SNE compress high-D into 2D/3D for visualization |

---

## 🧠 Bonus: **What Happens Beyond 3D? (nD)**

* **nD (n-Dimensional)** space is used in data science
* Each feature (input variable) is a dimension
* You can’t visualize beyond 3D easily — so we use **dimensionality reduction**

### Example:

* A dataset with 6 features (age, salary, city, education, experience, rating) lives in **6D space**
* We use **PCA** to compress into **2D or 3D** for graphs

---

| Concept      | 2D                            | 3D                             |
| ------------ | ----------------------------- | ------------------------------ |
| Axes         | x, y                          | x, y, z                        |
| Point Format | (x, y)                        | (x, y, z)                      |
| Use Case     | Plots, Regression             | Object detection, AR, robotics |
| In AI        | Feature spaces, visualization | Lidar, 3D CNN, GANs            |
